# Нанопроволоки и наностержни

# Конспект лекции

## Нановискеры

Возможность образования кристаллов в форме нитей в природе и технике известна давно. Очень длинные и тонкие нити карбоната кальция и гипса, которые можно завязывать в узел, были найдены в пещере французским спелеологом Кастером. Во время Второй мировой войны было обнаружено, что выход из строя радиодеталей и телефонной связи происходит из-за образования на поверхности гальванических покрытий тончайших металлических игл. Эти нити, диаметром порядка 1 мкм и длиной в $10^3-10^4$ раз больше, называют нитевидными кристаллами или усами, вискерами.

Первая работа о прочности нитевидных кристаллов олова, опубликованная американскими учеными в 1952 г., произвела сенсацию во всем мире. Прочность этих кристаллов превосходила в сотни раз прочность макроскопических образцов. Понятно, что сразу после этой публикации в материаловедческих  лабораториях мира началась разработка методов получения "усов" различных материалов и исследование их физико-химических свойств. К настоящему времени синтезированы  нитевидные кристаллы практически всех металлов, некоторых сплавов и неорганических соединений. Все они обладают гораздо большей прочностью по сравнению с моно- и поликристаллами этих же веществ.

Уменьшение в несколько раз электросопротивления металлических нитевидных 
кристаллов по сравнению с проволокой такого же диаметра также связано с отсутствием 
дефектов. Однако уменьшение диаметра до размера, когда длина пробега электронов проводимости становится сравнимой с ним, приводит к  увеличению сопротивления, поскольку электроны начинают уже взаимодействовать с поверхностью кристаллика. Высокое внутреннее совершенство в нитевидных кристаллах  диэлектриков приводит к увеличению в несколько раз электрического напряжения, при котором в обычных кристаллах происходит пробой диэлектрика.

При малых толщинах нитевидные кристаллы ферромагнетиков, как и наночастицы, могут
находится в однодоменном состоянии. В этом случае для перемагничивания (замена северного полюса на южный) нитевидного кристалла необходимо приложить магнитное поле в 1000 раз большее, чем для обычных кристаллов.

С совершенством строения поверхности наиболее тонких нитевидных кристаллов 
связана и их повышенная химическая и термическая устойчивость по сравнению со своими же, но более толстыми собратьями. Выращенные в условиях высокого вакуума нитевидные  кристаллы железа и кобальта диаметром менее 70 нм не окисляются на воздухе в течение  достаточно долгого времени. Чтобы нанести электролитическим методом на их поверхность слой атомов этого же металла, надо повысить напряжение на электродах в десятки раз. Оказывается, и испарить такие кристаллы гораздо труднее, чем "плохие", имеющие дефектную поверхность. Затраты энергии будут больше на несколько десятков процентов.

Нитевидные кристаллы обладают не только термической, химической, но и фазовой 
устойчивостью. Многие вещества могут существовать в нескольких модификациях (фазах) и переходят из одной в другую при заданном давлении только при определенной температуре. А вот в совершенных нитевидных кристаллах, чтобы произошел переход из фазы 1 (устойчивой при низкой температуре), в фазу 2 (устойчивую при высокой температуре) надо перегреть кристалл на несколько сот градусов. В этом опять "виноваты" дефекты, а точнее их отсутствие, поскольку при таком фазовом переходе в кристалле необходимо образование зародыша новой фазы, которое происходит 
гораздо легче в дефектном кристалле. 

Приведенные здесь "особые свойства" в большинстве случаев относятся только к 
химически чистым веществам, не содержащим атомов других веществ (примесей). Но если они туда попадают, то свойства нитевидных кристаллов могут резко измениться. Многие примеси сильно влияют на совершенство строения кристалла, что и приводит к изменению его свойств.

Перечисленные выше физико-химические и механические свойства нитевидных кристаллов нашли применение для создания композиционных материалов, где они используются в качестве наполнителей для повышения прочности, особенно  жаропрочности конструкций, работающих при высоких и очень высоких температурах (сопла ракет, теплозащита). Отдельные нитевидные кристаллы используются как высокоточные датчики для измерений температуры и давления, а также в электронной технике.

## Нанопроволоки

Нанопроволоки — одномерные, анизотропные структуры, характеризующиеся малыми размерами в диаметре и большим отношением площади поверхности к объему. В отличие от других низкоразмерных систем, у нанопроволок есть два квантово-ограниченных направления и одно неограниченное, доступное для электрической проводимости. Это позволяет использовать нанопроволоки там, где требуется не перенос заряда посредством тунелирования, а электрическая проводимость. 

Основными методами синтеза являются:
- Синтез с использованием опорного трафарета
    - Нагнетение под давлением
    - Электрохимическое осаждение
    - Осаждение из газовой фазы
- Метод синтеза пар-жидкость-кристалл

Синтез с использованием опорного трафарета

Синтез с использованием трафарета — концептуально простой и наглядный способ формирования наноструктур. Трафарет (шаблон) представляет собой пластину с множеством малых цилиндрических пор или полостей, которые заполняются необходимым материалом, принимающим форму этих пустот, в результате чего получаются нанопроволоки. В качестве трафарета для синтеза нанопроволок используются пластины из анодированного оксида алюминия Al$_2$O$_3$; стекло с наноканалами; полимеры и пленки слюды, поры в которых формируются травлением трека, оставленным при бомбардировке ионным пучком. Пористые анодированные пластины оксида алюминия формируются электрохимическим травлением Al фольги или пленок в щавелевой и прочих кислотах. При тщательно выбранных условиях анодирования пленка оксида алюминия обладает регулярной гексагональной структурой из почти цилиндрических каналов. Регулярная структура образуется в результате тонкого баланса между ускоренной электрическим полем диффузией реагентов и растворением оксида алюминия. Одна из гипотез, объясняющих процесс самоупорядочения пор, основывается на идее механических напряжений на интерфейсе алюминий-оксид в результате расширения материала при анодировании. В зависимости от условий анодирования диаметр пор можно изменять от 10 до 200 нм с плотностью пор 10$^9$-10$^11$ см$^{-2}$. Как было показано многими исследовательскими группами, распределение размера пор и их упорядочение могут быть значительно улучшены с использованием двух стадий анодирования, когда пленка оксида алюминия растворяется после первого анодирования в кислотном растворе, с последующими повторным анодированием при тех же условиях.

Другой тип пористых материалов с упорядоченной структурой пор — стекло с наноканалами с плотностью упаковки пор порядка $10^{10}$ см$^{-2}$ и диаметром пор 2-10 нм. Еще один тип пористого трафарета, применяемый для синтеза нанопроволок, изготавливается путем химического травления мембран из слюды или поликарбоната, подвергнутых действию ионных пучков.

** Получение нанопроволок трафаретным методом с использованием нагнетения под давлением **

Процесс нагнетения под давлением часто используется для получения кристаллических нанопроволок из материалов с низкой температурой плавления, используя пористые пластины с достаточной механической прочностью. В этом методе проволоки формируются нагнетением под давлением необходимого материала в жидкой форме в вакуумированные поры пластины. Из-за нагревания и внешнего давления трафарет,, используемый в этом методе, должен быть химически устойчивым и не разрушаться при температуре и давлении процесса. Пленки por-Al$_2$O$_3$  и стекло с наноканалами — два типичных материала, используемых в качестве трафаретов. Так были получены металлические (Bi, Sn, Al) и полупроводниковые (Si, Te, GaSb, Bi2Te3) нанопроволоки.

Давление $P$, требуемое, для того чтобы преодолеть поверхностное натяжение жидкого материала и наполнить поры диаметра $D$, определяются уравнением Вашбурна:

$$
D = -\dfrac{4 \gamma cos \theta}{P},
$$

где $\gamma$ — поверхностное натяжение жидкости и $\theta$ краевой угол между жидкостью и пластиной. Чтобы уменьшить требуемое давление и увеличить коэффициент заполнения, можно использовать некоторые поверхностно-активные вещества с целью уменьшения поверхностного натяжения и краевого угла смачивания. Кроме того, таким же эффектом могут обладать специально вводимые в нагнетаемый материал компоненты:. Так, введение Cu в расплав Bi упрощает задачу заполнения пор в por-Al$_2$O$_3$ жидким висмутом, и может увеличить число сформированных нанопроволок. Однако некоторые из поверхностно-активных веществ могут вызывать проблемы, связанные с загрязнением, и их следует использовать с осторожностью. 

Нанопроволоки, сформированные методом нагнетания под давлением обычно обладают высокой степенью кристалличности и предпочтительной ориентацией кристалла вдоль оси проволоки. Кристаллографическая ориентация может зависеть от диаметра полостей используемого шаблона, а также от применяемого давления. При этом наиболее плотно упаковонное направление вдоль оси наблюдается при более высоких давлениях нагнетания.

**Электрохимическое осаждение**

Электрохимическое осаждение происходит по нормали по нормали к поверхности подложки, его можно применять для формирования одномерных и нульмерных наноструктур при ограничении осаждения в пределах пор соответствующего трафарета. Сначала на одну из сторон пористой мембраны наносится тонкая проводящая пленка, которая служит катодом при дальнейшем осаждении. Длину осажденных проволок можно контролировать, изменяя время электроосаждения. Выбранный трафарет должен бытьхимически устойчивым в электролите во время процесса электролиза. При необходимости в процессе осаждения используются поверхностно-активные вещества, в частности, спирты. Использование ПАВ способствует заполнению большего числа пор.

В случае осаждения в поры por-Al$_2$O$_3$ обычно химическим травлением удаляется изолирующий барьер с обратной стороны. Однако при использовании вместо постоянного напряжения переменного можно использовать выпрямляющие свойства оксидного барьера и провожить электролитическое осаждение на переменном токе. В таком случае при использовании синусоидальных сигналов несмотря на симметричность гармонического сигнала, величина тока больше в течение катодных полупериодов, рпи этом происходит преобладание процесса осаждения над процессос растворения, происходящим в течение анодных полупериодов. Так как в дефектных местах не происходит процесса выпрямления и процесса осаждения и растворения уравновешены, в этих местах материал не осаждается. 

В отличие от нанопроволок, синтезируемых методом нагнетания под давлением, нанопроволоки, изготовленные электрохимическим осаждением, обычно являются поликристаллическими, без предпочтительных ориентаций кристаллов, что доказано исследованием с помощью рентгеновской дифракции. Однако существуют некоторые исключения. Например, у поликристаллических пленок сульфида кадмия CdS, изготовленных методом электроосаждения на переменном токе, существуют доминирующая ориентация роста вдоль $c$-оси. Для CdS, CdSe, CdTe также были получены и монокристаллические нанопроволоки при использовании неводного электролита. Показано, что в ряде случаев можно получить монокристаллические нанопроволоки при использовании импульсного напряжения, так как концентрация ионов металла в приповерхностной области раствора выравнивается в промежутках между импульсами.

Одно из преимуществ метода электрохимического осаждения — воможность изготовления многослойных нанопроволочных структур. Изменяя катодные потенциалы в электролите, содержащем два вида различных ионов можно попеременно контролируемо осадить два вида металла. Так, например, были получены нанопроволоки Co/Cu диаметром 40 нм. Таким образом, метод электролитического осаждения — доступный способ формирования многослоынйх одномерных наноструктур.

**Осаждение из газовой фазы** 

Осаждение нанопроволок из газовой фазы включает термовакуумное осаждение из паровой фазы, химическое осаждение из газовой фазы (CVD), химическое осаждение из газовой фазы разложением металлорганических соединений (MOCVD). Как и электрохимическое осаждение, осаждение из газовой фазы дает возможность формирование нанопроволок меньшего диаметра, чем методами нагнетания давлением, в этом случа нет необходимости прикладывать высокое давление для преодоления силы поверхностного натяжения при заполнении пор. 

**Анизотропный рост нанопроволоко по методу пар-жидкость-кристалл**

Описанный далее процесс был вначале предложен для выращивания нитевидных монокристаллов кремния от 100 нм до сотен микро в диаметре. Предложенный механизм выращивания включает поглощение источника материала из газовой фазы в жидкую каплю катализатора (расплавленной частицы золота). В расплаве происходит пересыщение, зародышеообразование и осаждение твердого материала. Этот зародыш кристаллизации служит предпочтительным центром захвата при дальнейшем осаждении материала на поверхности раздела жидкости. В результате происходит рост зародыша кристаллизации в нитевидный кристалл, при этом подавляются другие возможные процессы зародышеобразования. Так как капля жидкого металла катализурует включение материала из газового источника в растущий кристалл, осажденный слой растет анизотропно в виде нити, диаметр которой определяется диаметром капли металла. Нанопроволоки, полученные таким образом имеют высокую чистоту, за исключением конца, который содержит отвердевшую частицу катализатора. Для уменьшения среднего диаметра проволоки до нанометрового размера необходимо формирование наноразмерных капель катализатора. 

**Синтез в коллоидных растворах**

Чтобы поддержать анизтропный одномерный рост кристаллов в растворах необходимо использовать ПАВ. Однако при коллоидном синтезе обычно получаются не проволоки, а наностержни (наночастицы с аспектным отношением ~10). Контроля над формой частиц можно добиться за счет применения смесей ПАВ различной природы, избирательно адсорбирующихся на гранях различной кристаллографической ориентации. 

## Упорядочение нанопроволок в структуры

Применяют следующие методы совмещения и ориентации нанопроволок после синтеза:
- на основе микрофлюидных каналов;
- сборка методом Лэнгмюра-Блоджетт;
- сборка с использованием электрического поля.

Метод микрофлюидных каналов заключается в ориентации нанопроволок направлением потока жидкости, когда раствор нанопроволок вводится в систему микрофлюидных каналов. Второй метод основан на упорядочивании нанопроволок на границах жидкость-газ или жидкость-жидкость с использованием сжимающих напряжений на поверхности раздела. Выровненные пленки нанопроволок могут быть перенесены на подложку, а методы литографии использованы для создания межсоединений. Используя этот метод, можно формировать массивы нанопроволок размерами порядка сантиметров.

Третья методика основанв на явлении электрофореза, в результате которого поляризованные нанопроволоки собираются в областях с высокой напряженностью электрического поля. Нанопроволоки упорядочиваются между двумя электродами, на которые подается переменное наряжение.

Наиболее практичным с позиций микроэлектроники является подход, в котором наноструктуры упорядочиваются непосредственно в процессе их роста за счет заранее определенного шаблона, полученного с применением фото- или электронно-лучевой литографии. Такие шаблоны позволяют получать массивы столбиков нанопроволок в предопределенных местах, вертикально упорядоченных и с одинаковой ориентацией выращенных кристаллов. 

## Квантовые нити

Рассмотрим тонкую нить полупроводника, в которой движение электронов резко ограничено в двух направлениях из трех и лишь вдоль оси нити (будем называть ее осью $x$) остается свободным. При этом за счет малых поперечных размеров нити движение в плоскости $yz$ квантуется, и его энергия может принимать лишь  дискретные значения $E_i$, так что полный закон дисперсии имеет вид

$$E = E_i + \frac{p^2_x}{2m},$$

где $m$ — эффективная масса электронов. Видно, что каждому дискретному уровню $E_i$ соответствует набор возможных состояний отличающихся импульсом $p_x$. При этом обычно говорят не об уровне, а о подзоне размерного квантования с номером $i$.

### Методы изготовления квантовых нитей

Большинство способов изготовления квантовых нитей основывается на том, что в системе с двумерным электронным газом (как правило, на основе гетероструктур) тем или иным способом ограничивается движение электронов еще в одном из направлений. Для этого есть несколько способов.

Наиболее очевидный из них – это непосредственное “вырезание” узкой полоски с помощью литографической техники. При этом для получения электронных нитей шириной в сотни ангстрем, где квантование энергий электронов будет заметным, необязательно делать полоски имен но такой ширины, что требует литографической техники сверхвысокого разрешения. Дело в том, что на боковых гранях вытравленной полоски, как и на свободной поверхности полупроводника, образуются поверхностные состояния, создающие, как правило, слой обеднения. Этот слой вызывает дополнительное сужение проводящего канала, в результате чего квантовые эффекты можно наблюдать и в полосках большей ширины – порядка десятой доли мкм.

Можно поступить и иначе. Поверхность полупроводниковой структуры покрывают металлическим электродом, создающим с полупроводником контакт Шоттки и имеющим узкую щель. Если гетерограница находится достаточно близко от поверхности, в слое обеднения, то двумерные электроны на границе будут отсутствовать всюду, кроме узкой области под щелью. Такой тип одномерной структуры обладает дополнительным преимуществом: меняя напряжение на затворе, мы можем управлять эффективной шириной квантовой нити и концентрацией носителей в ней.

# Прочность

Одной из количественных характеристик прочности материалов служит предельное напряжение ($p_c$, Па), при котором происходит разрыв образца (стержня) при одноосном растяжении. предельное напряжение определяется уравнением

$$p_c = \dfrac{F_c}{S},$$

где $F_c$ — растягивающая сила, вызывающая разрыв, $S$ — площадь поперечного сечения образца. Для образцов с достаточно большим поперечным сечением (диаметр > 100 мкм) прочность зависит только от химической природы вещества. Однако для тонких образцов, диаметр которых соответствуют размеру дисперсных частиц показан явный размерный эффект. Он заключается в том, что предельное напряжение увеличивается по мере уменьшения диаметра стержней, волокон, частиц и т.д. Этот масштабный эффект иллюстрирует зависимость прочности стеклянных нитей от их диаметра

|$d$, мкм | $p_c$, Па|
|:-:|:-:|
|22,0 | 220  |
|16,0 | 1070 |
|12,5 | 1460 |
|8,0  | 2070 |
|2,5  | 5600 |

Приведенные данные показывают, что прочность тонких нитей резко возрастает по мере уменьшения их диаметра. Предельное напряжение зависит не только от природы вещества, но и от различных дефектов структуры. Наиболее сильное влияние на механические свойства оказывают так называемые "линейные" дефекты — краевые и винтовые дислокации. Вероятность нахождения в образце "опасного" дефекта, ответственного за его разрушение, тем ниже, чем меньше диаметр образца. Среднее расстояние между дислокациями в кристаллах составляет примерно 100 нм. Поэтом рост прочности характерен для образцов меньших размеров. Это положение легло в основу получения высокопрочных композитных материалов на основе нитевидных волокон.
