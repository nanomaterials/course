# Конспект лекции

Общие свойства материалов

Широкозонные полупроводники и диэлектрики не поглощают излучение в видимом диапазоне. Однако в случае наночастиц возможно наблюдение рассеяния света. В частицах мала концентрация свободных носителей заряда, однако сопротивление образцов за счет развитой поверхности может меняться в результате внешнего воздействия (сенсорика). Малая проводимость также интересна для использования пьезоэлектрических свойств. Химическая инертность представляет интерес для создания покрытий и оболочек других частиц.

### Гетерогенный катализ

процессы адсорбции различных веществ на поверхности (первая химическая стадия в гетерогенном катализе) стали изучать на молекулярном уровне многочисленными физическими методами. органических реакций) химии. На поверхности металлов образуются первичные комплексы реагентов с атомом или группой атомов поверхности и продукты их превращений.

## Фотокаталитические методы очистки воды и воздуха

*Фотокатализ*  — изменение скорости или возбуждение химических реакций под действием света в присутствии веществ (фотокатализаторов), которые поглощают кванты света и участвуют в химических превращениях участников реакции, многократно вступая с ними в промежуточные взаимодействия и регенерируя свой химический состав после каждого цикла таких взаимодействий.

Рассмотрим определение на примере реакции дегидрирования этилового спирта в водных растворах

C$_2$H$_5$OH $\rightarrow$ C$_2$H$_4$O + H$_2$ $\uparrow$

При комнатной температуре эта реакция не идет. Она может идти под действием света с длиной волны $\lambda < 205$ нм, которая соответствует краю поглощения этанола. Но это жесткий ультрафиолет, который практически отсутствует в солнечном спектре, достигающем поверхности Земли. В присутствии мелкодисперсного TiO$_2$ эта реакция идет с высоким квантовым выходом под действием света c $\lambda < 400$ нм. Фотонов с такими длинами волн много в солнечном спектре. Фотокатализаторы возбуждают химическую реакцию или, как говорят, расширяют спектр действия системы, то есть область длин волн света, при облучении которыми идет реакция.

При воздействии соответствующего излучения в полупроводниковых частицах происходят следующие процессы:
а) поглощение электромагнитного излучения с рождением электронно-дырочных пар,
б) диффузия электронов и дырок к поверхности полупроводника,
в) объемная рекомбинация электронов и дырок,
г) поверхностная рекомбинация электронов и дырок,
д) полезные реакции электронов и дырок с адсорбированными молекулами.

Схематически процессы, происходящие на частице TiO$_2$, изображены на рисунке.

![](Figs/TiO2_NP.png)

*Рис. Схематическое изображение процессов, идущих на полупроводниковой частице* [Савинов, 2000]

В терминах окислительно-восстановительных потенциалов реакционная способность электрона и дырки на поверхности TiO2 характеризуется следующими величинами: потенциал электрона ∼ −0,1 В, потенциал дырки ∼ +3 В относительно нормального водородного электрода. Инымисловами, электрон способен реагировать с кислородом, рождая последовательность реакций:

- $e^-$ + O$_2$ $\rightarrow$ O$_2^-$,
- O$_2^-$ + $e^-$ $\rightarrow$ O$_2^{2-}$ $\rightarrow$  O$^-$ + O$^-$,
- O$_2^{2-}$ + 2H$^+$ $\rightarrow$ H$_2$O$_2$,
- O$^-$ + $e^-$ $\rightarrow$ O$^{2-}$
- H$_2$O$_2$ + $e^-$ $\rightarrow$ OH$\cdot$ + OH$^-$,
- O$^-$ + H$^+$ $\rightarrow$ OH$\cdot$,
- $h^+$ + H$_2$O $\rightarrow$ OH$\cdot$ + H$^+$,
- $h^+$ + C$_x$H$_y$O$_z$  $\rightarrow$ C$_x$H$_{y-1}$O$_z$ + H$^+$

Образующиеся молекулы такие как O$^-$ и радикал OH$\cdot$ являются мощными окислителями. Таким образом, поверхность наночастиц становится под светом сильнейшим окислителем.

Квантовый выход реакции определяется как отношение числа носителей заряда, участвующих в фотокаталитической реакции к числу поглощенных квантов электромагнитного излучения. Квантовый выход реакции можно представить следующим образом:

$$\Phi = \eta_i \eta_r,$$

где $\eta_i$ — доля носителей тока, достигших поверхности, $\eta_r$ доля носителей тока, вступивших в полезную реакцию (избежавших поверхностной рекомбинации).

Если размер частицы сравним или меньше длины свободного пробега носителей заряда, то $\eta_i \rightarrow 1$.

Из эксперимента известно, что для частиц TiO$_2$ с диаметром 5 нм все носители заряда выходят на поверхность. Однако на практике не всегда самыми фотоактивными являются порошки с малыми частицами. Это связано со второй составляющей квантового выхода реакции:

$$\eta_r = \dfrac{V_r}{V_{sr} + V_r},$$

где $V_{sr}$ — скорость поверхностной рекомбинации, $V_r$ — скорость полезной реакции. Обе скорости могут заметно изменяться в зависимости от структуры поверхности, то есть от строения и энергетики центров адсорбции и захвата носителей тока. Из практики известно, что наибольшей фотокаталитической активностью обладают образцы TiO$_2$ с кристаллической модификацией анатаза и не содержащие большого числа примесей.

Показано, что на поверхности TiO$_2$ могут быть окислены (минерализованы) до CO$_2$ и H$_2$O практически любые органические соединения. Если в состав соединений входят азот или атомы галогена Х, то в продуктах реакции будут наблюдаться HNO$_3$ и HX. Единственным известным примером соединения, которое не подвергается на поверхности TiO$_2$ окислению под действием света, является тетрахлорметан.

На практике любой фотокаталитический очиститель воздуха включает в себя пористый носитель с нанесенным TiO$_2$, который облучается светом и через который продувается воздух. Так, на рисунке показано
