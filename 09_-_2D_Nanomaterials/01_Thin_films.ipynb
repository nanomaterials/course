# Конспект лекции

## Тонкие пленки

Изменение свойств приповерхностных слоев твердых тел с целью придания им заданных свойств, таких, как твердость, износостойкость, коррозионная стойкость и др., давно является предметом сознательной деятельности материаловедов. Достигается это изменением состава или структуры приповерхностного слоя твердого тела. К числу технических приемов,  приводящих к такому изменению, относятся различные способы химико-термической обработки (азотирование, поверхностная закалка и т.п.). Однако в традиционных схемах таких процессов, в основе которых лежит диффузионный перенос атомов или тепла, толщина слоев измененного химического состава или структуры достигает величины в несколько миллиметров. О таких материалах трудно говорить как о тонкопленочных. Современные способы химико-термической обработки материалов в отличие от традиционных, использующих 
нагревательные печи, базируются на лазерной, электронно- или ионно-лучевой обработке поверхности материалов. В этом случае за счет особенностей взаимодействия этих энергоносителей с твердыми телами изменение свойств твердых тел локализуется в чрезвычайно тонких приповерхностных слоях твердых тел, т.е. с их помощью создаются объекты, которые уверенно могут быть отнесены к тонким пленкам.

## Поверхностные наноструктуры

Под терминами “поверхность” или “межфазная граница” обычно понимают слой вещества конечной толщины, разделяющий различные объемные фазы. Его толщина определяется теми конкретными свойствами, которые изучаются и составляет, как правило, несколько атомных слоев вещества. Несмотря на то что при анализе поверхностных явлений обычно используют модель однородной поверхности, даже в случае наиболее совершенных кристаллов поверхность оказывается
сильно неоднородной по структурным, электрическим и химическим свойствам. ## Тонкая пленка как двумерная электронная система

Пусть имеется тонкая металлическая или полупроводниковая пленка толщиной $а$. То обстоятельство, что в обычных условиях электроны проводимости сосредоточены в пленке и не выходят из нее в окружающую среду, означает, что материал пленки представляет собой потенциальную яму для электронов с глубиной, равной работе выхода $\Phi$, и шириной $a$. Согласно законам квантовой механики, энергия электронов в такой яме может принимать лишь некоторые дискретные значения $E_n$, называемые уровнями размерного квантования. Поскольку типичное значение $\Phi$  4-5 эВ на несколько порядков превосходит характерную тепловую энергию электронов $kT$, равную при комнатной температуре 0,026 эВ, то стенки ямы можно считать бесконечно высокими. В этом случае значения уровней энергии даются формулой:

$$
E_n = \dfrac{n^2 \hbar^2}{2m_ea^2},
$$

где $m_e$ — эффективная масса электрона. Если потенциальная яма, ограничивающая движение электронов, имеет конечную глубину и, вообще говоря, не прямоугольную форму, то формула неприменима, однако сам вывод о дискретности энергетического спектра остается в силе. Сохраняет силу и порядковая оценка для энергий квантовых уровней, вытекающая из гейзенберговского отношения неопределенностей:

$$
E_n \sim \dfrac{\hbar^2}{m_ea^2}.
$$

Приведенные формулы для $E_n$ относятся лишь к движению поперек потенциальной ямы
(по оси $z$). На движение носителей в плоскости $ху$ (параллельно стенкам ямы) потенциал ямы не влияет. В этой плоскости носители движутся как свободные. Полная же энергия квантово-размерной системы носит смешанный дискретно-непрерывный характер, представляя собой сумму энергии дискретных уровней, связанных с движением в направлении квантования, и непрерывной компоненты, описывающей движение в плоскости слоя:

$$
E = E_n + \dfrac{p_x^2 + p_y^2}{2m_e},
$$

где $p_x$ и $p_y$ — соответствующие компоненты импульса. Подобное изменение спектра за счет размерного квантования приводит к существенному изменению всех электронных свойств системы по сравнению с массивным образцом .

Если концентрация электронов в пленке не слишком велика, а тепловая энергия много меньше расстояния между уровнями ($E_2 - E_1 >> kT$), то подавляющая часть носителей находится на первом уровне. Тогда никакой упругий процесс (например, рассеяние на примесях или акустических фононах) не может изменить квантовое число $n$, переводя электрон на вышележащий уровень, так как это потребовало бы дополнительных затрат энергии. Это означает, что электроны при упругом рассеянии могут изменять только свой импульс в плоскости $ху$ то есть ведут себя как чисто двумерные частицы. Поэтому квантоворазмерные структуры, в которых заполнен лишь один квантовый уровень, называют структурами с двумерным электронным газом, несмотря на конечность их размеров в направлении оси $z$.

Интенсивность рассеяния обычно характеризуется средним временем $\tau$ между двумя последовательными актами рассеяния, связанным прямой пропорциональностью с другой важной характеристикой носителей — их подвижностью $\mu = e\tau/m$. Величина $\tau$ представляет собой среднее время жизни носителей
в состоянии с данным импульсом. В силу соотношения неопределенностей конечная величина $\tau$ влечет за собой неопределенность в энергии данного состояния $\Delta E \ge \hbar/\tau$. Говорить о наличии в системе отдельных дискретных уровней можно лишь в случае, когда расстояние между ними превышает неопределенность $\Delta E$, то есть при условии

$$
E_2 - E_1 >> \dfrac{\hbar}{\tau} = \dfrac{\hbar e}{m_e \mu_e}.
$$

Можно показать, что последнее условие эквивалентно требованию того, чтобы длина
свободного пробега электронов $l$ значительно превосходит толщину пленки.  Согласно квантовой механике, квантование возникает, когда частица совершает периодическое движение. Это происходит лишь при достаточно слабом рассеянии, когда электрон между двумя актами рассеяния (то есть пройдя путь длиной $l$) успевает совершить несколько периодов колебаний (несколько раз пересечь пленку по толщине).

Поскольку расстояние между уровнями размерного квантования пропорционально $a^{-2}$, то из условий размерного квантования следует, что для наблюдения квантовых размерных эффектов необходимы малые толщины слоев, высокие подвижности носителей и достаточно низкие температуры. В частности, чтобы наблюдать квантовые размерные эффекты в наиболее распространенных полупроводниках при температурах вплоть до комнатной, необходимо, чтобы $a \le 10$нм, а $\mu_e$ превосходило несколько тысяч $\text{см$^2$/(В$\cdot$с)}$.

Для тонких пленок существует еще одно условие, необходимое для наличия квантования, — это высокое качество поверхности, нужное для зеркального отражения носителей от границы пленки. Если это не так, то при каждом отражении частица "забывает" о своем состоянии до отражения, то есть на границе происходит рассеяние, разрушающее квантование. Для реализации зеркального отражения на границах необходимо, чтобы размеры шероховатостей на поверхности были меньше де-бройлевской длины волны носителей. Кроме того, поверхность пленки не должна содержать высокой плотности примесей, приводящих к дополнительному рассеянию.

## Свойства структур с двумерным электронным газом

Опишем оптические характерстики структур с двумерными электронами. Главная особенность здесь — та же, что и в квантовых точках — возможность управления эффективной шириной запрещенной зоны. Минимальная энергия, которую может иметь электрон равна $E_{c1}$ (энергия отсчитывается от дна зоны проводимости), а минимальная энергия дырки — $E_{v1}$ (если отсчитывать энергию от потолка валентной зоны). Поэтому край оптического поглощения структур с квантовыми ямами определяется уже не шириной запрещенной зоны $E_g$, а суммой $E_g + E_{c1} + E_{v1}$. Энергии квантования $E_{c1}$, $E_{v1}$ зависят от ширины квантовой ямы $a$, поэтому меняя толщину слоев в гетероструктуре, можно из одних и тех же полупроводников изготавливать приборы оптоэлектроники (лазеры, светодиоды, фотоприемники) для различных областей спектра.

Более того, для уже изготовленной квантовой ямы с фиксированной шириной край поглощения можно менять, прикладывая к
структуре электрическое поле, перпендикулярное слоям. При этом меняется форма квантовой ямы, энергии уровней $E_{c1}$, $E_{v1}$ и положение края поглощения. В результате для света с частотой, близкой к краю поглощения для данной квантовой ямы, коэффициент поглощения, а следовательно и интенсивность света, прошедшего через структуру, может
сильно меняться под действием электрического поля. Поскольку электрическое поле можно менять с очень высокой частотой, то это открывает возможности высокочастотной модуляции света, то есть наложения на него огромного объема информации для передачи по системам
оптической связи, которая сейчас повсеместно вытесняет связь с помощью проводов и радиоволн.

Не менее важную роль играют структуры с квантовыми ямами в создании современных полупроводниковых лазеров. Помимо уже упоминавшейся возможности перестройки частоты, они обладают еще одним важным преимуществом. Неравновесные носители, создаваемые за счет инжекции или освещения в достаточно широкой области широкозонного материала,
захватываются в узкозонный слой, образующий квантовую яму, где их потенциальнаяэнергия меньше. За счет малой толщины $а$ этого слоя объемная плотность неравновесных носителей в нем очень велика и легко достигает значений, необходимых для лазерной генерации, уже при достаточно слабой накачке. Использование структур с квантовыми ямами позволило снизить во много раз пороговый ток инжекционных лазеров (тот минимальный ток, который надо пропустить через прибор, чтобы получить лазерное излучение) и приступить к промышленному производству миниатюрных и надежных лазеров, дающих непрерывное излучение при комнатной температуре.

Другой важной особенностью структур с двумерными электронами является наличие у них сильного инфракрасного поглощения за счет электронных переходов между уровнями размерного квантования. Это дает возможность
создавать высокоэффективные приемники для такого излучения. Принцип действия такого приемника показан на рис. На нем изображен ход края зоны проводимости в многослойной гетероструктуре, содержащей
много квантовых ям (ход валентной зоны не показан, поскольку нас будут интересовать только переходы электронов внутри зоны проводимости). Если такая структура легирована донорной примесью, то в равновесии все электроны будут сосредоточены в квантовых ямах. Поэтому в направлении, перпендикулярном слоям (ось $z$), сопротивление структуры очень
велико. Инфракрасный свет с энергией кванта, превосходящей глубину залегания уровня $\delta_Е$, вызывает переходы электронов двумерного
газа в делокализованные состояния над ямой, способные проводить ток в $z$-направлении, что понижает сопротивление структуры на величину, пропорциональную интенсивности света.

Описанную идею удалось применить для создания на основе квантовых ям с двумерным электронным газом не только одиночных фотоприемников, но и целых фоточувствительных матриц, обеспечивающих передачу телевизионного изображения в инфракрасном свете с длиной волны порядка 10 мкм. Такой свет излучается сравнительно слабо нагретыми телами (например, человеческое тело, машина или самолет с работающим двигателем и др.). Таким образом, речь идет о телевизионных
приемниках, позволяющих видеть в темноте указанные предметы, что имеет очень широкое применение в промышленности, медицине, военной технике и других областях.

В двумерных наноструктурах меняется также и энергетический спектр экситона:

$$
E_n^{2D} = \dfrac{E_1^{3D}}{\left(n-\frac{1}{2}\right)^2},
$$

где $E_1^{3D}$ — энергия основного состояния экситона в объемном полупроводнике, $n$ — главное квантовое число. Можно видеть, что уже при $n = 1$ энергия связи 2D-экситона увеличивается в 4 раза. Эффективный радиус экситона уменьшается вдвое по сравнению с трехмерным случаем.

## Свойства различных тонких пленок

**Алмазные пленки**. Следуя природе и законам термодинамики, люди в тех же условиях научились создавать искусственные алмазы. Оказалось, что алмаз может быть создан и при низких давлениях и умеренных температурах. Это удается осуществить в тех случаях, когда алмаз получается в виде тонких пленок. Сегодня газофазными методами успешно выращиваются эпитаксиальные монокристаллические и поликристаллические алмазные пленки, а также алмазоподобные пленки. Если алмазные пленки имеют кристаллическую структуру алмаза, то *алмазоподобные пленки* — это аморфные углеродные пленки с плотностью, близкой к плотности алмаза.

Алмазные пленки характеризуются теми же свойствами, что и массивный алмаз, т. е. это самый твердый и износостойкий из известных сегодня материалов. Алмаз сочетает в себе два практически несовместимых свойства: он — диэлектрик, т. е. не проводит электрический ток, но он же хорошо проводит тепло.

**Мембраны**. Современная технология микроэлектроники позволяет создавать тонкие пленки толщиной порядка одного микрона и менее. Примечательным оказывается то, что эти пленки не лежат на подложке, а являются фактически мембранами и соприкасаются с воздухом обеими своими поверхностями. Такие мембраны составляют сердцевину  широкого класса так называемых микросистем — миниатюрных датчиков температуры, давления, ускорений, газоанализаторов и т. п. На рис. представлена принципиальная схема изготовления мембраны из нитрида кремния, включающая стадии осаждения нитрида кремния, анизотропного травления и формирования структуры элемента будущего прибора в мембране или на ней. Миниатюрные датчики, изготовленные на базе таких мембран, кроме высокой чувствительности, характеризуются еще и тем, что относятся к разряду интеллектуальных датчиков, которые преобразуют сигнал датчика в электрический, обрабатывают его и выдают необходимые указания.

## Теоретические основы роста тонких пленок

При осаждении тонких пленок преобладают гетерогенные процессы, включающие в себя гетерогенные химические реакции, испарение, адсорбция и десорбцию на ростовых поверхностях, гетерогенную нуклеацию и рост поверхности. Большинство процессов выращивания пленок и определение их параметров проводятся в вакууме.

Рост тонких пленок, как и все фазовые переходы, включает в себя процессы нуклеации и роста на подложке и поверхностях роста. Процесс нуклеации играет ключевую роль в задании кристалличности и микроструктуры образующихся пленок. Процесс нуклеации особенно важен для наращивания тонких пленок с нанометровой толщиной. Чрезвычайно важно учитывать взаимодействие между пленкой и подложкой. Многочисленные экспериментальные наблюдения выявили, что существуют [три основных режима нуклеации роста тонкой пленки](https://ru.wikipedia.org/wiki/Механизмы_роста_тонких_плёнок):

1. Островковый — [механизм Фольмера-Вебера](https://ru.wikipedia.org/wiki/Механизм_роста_Вольмера_—_Вебера),
2. Послойный — [механизм Франка-Ван дер Мерве](https://ru.wikipedia.org/wiki/Механизм_роста_Франка_—_Ван_дер_Мерве),
3. Промежуточный механизм роста, островково-послойный — [механизм Странского-Крыстанова](https://ru.wikipedia.org/wiki/Механизм_роста_Странского_—_Крыстанова).

*Островковый рост* имеет место при плохом смачивании, когда компоненты наращиваемого вещества сильнее связываются друг с другом, чем с подложкой. Этот тип нуклеации на начальном этапе формирования пленки демонстрируют многие металлы на изолирующих подложках, подложках из галогенидов щелочных металлов, графита и слюды. Последующий рост приводит к слиянию островков с образованием сплошной пленки.

![](Figs/mechanisms.png)
*Рис. Механизмы роста тонкого слоя на поверхности монокристалла: а – двумерный (послойный), б – трехмерный (островковый), в – промежуточный механизм роста (механизм Странского-Крастанова)*

Противоположностью островкому росту является *послойный рост*. В этом случае осаждаемый материал эффективно смачивает подложку, то есть ростовые элементы сильнее связываются с подложкой, чем друг с другом. Сплошной монослой формируется до того как начинается осаждение второго слоя. Так происходит эпитаксиальный рост монокристаллических пленок.

Островково-послойный рост — комбинация послойного и островкового роста. Такая разновидность обычно сопровождается напряжениями, возникающими во время формирования ядер или пленок, например, в результате отличия параметров решеток подложки и осаждаемого материала.

Наиболее важным случаем послойного роста является выращивание монокристаллических пленок посредством либо гомоэпитаксии, когда наращиваемая пленка имет ту же кристаллическую структуру и химический состав, что и подложка, либо гетероэпитаксии, когда кристаллическая струкутра наращивамой пленки близка к кристаллической структуре подложки. Гомоэпитаксия — это практически рост самой подложки, следовательно, между подложкой и наращиваемой пленок фактически нет границы раздела и процесс нуклеации отсутствует.

### Молекулярно-лучевая эпитаксия

Необходимые структуры получаются методом [молекулярно-лучевой эпитаксии](https://ru.wikipedia.org/wiki/Молекулярно-пучковая_эпитаксия) (МЛЭ). Процесс МЛЭ – это процесс испарения и конденсации вещества в сверхвысоком вакууме ($Р < 10^{-9}$ мм рт. ст.). Проводят его в специальной камере, схематическое изображение которой показано на рис. Наличие вакуумного шлюза позволяет менять подложки, сохраняя высокий вакуум. Для уменьшения давления остаточных газов вся свободная площадь камеры ограждается экранами, охлаждаемыми жидким азотом. Эффузионные (эффузия – медленное истечение газа через малое
отверстие) испарительные ячейки выполняются из тугоплавкого материала, например нитрида бора BN.

Для арсенида галлия ее температура должна составлять около 600°С, для кремния – не ниже 850°С. Процесс роста слоев начинается при открытии основной заслонки и заслонок соответствующих испарителей. Сверхвысокий вакуум и малая скорость поступления атомов на растущую поверхность (примерно $10^{14}-10^{15}$ атомов в секунду) приводят к эпитаксиальному росту пленок посредством практически монослойного заполнения растущей поверхности, обеспечивая исключительно точное управление профилями химического состава и легирования. В процессе МЛЭ возможен непосредственный контроль как состава газовой фазы (масс-спектрометрия), так и параметров слоев: кристаллической структуры (методами дифракции быстрых или медленных электронов), химического состава (оже-спектроскопия), толщины (эллипсометрия), датчики которых могут быть помещены в высоковакуумную камеру установки. Обладая неоспоримыми преимуществами, метод МЛЭ имеет единственный, но весьма существенный недостаток – высокую стоимость, что вызвало поиск альтернативных методов.

### Газофазная эпитаксия из металлоорганических соединений

Одним из методов альтернативных МЛЭ явился метод [газофазной эпитаксии](https://ru.wikipedia.org/wiki/Газофазная_эпитаксия) из металлоорганических соеди-
нений (ГФЭ МОС) или так называемая МОС-гидридная технология (особенно ее модификация при пониженных давлениях). В этом случае исходные газообразные реагенты пиролитически (под действием высокой температуры) разлагаются у поверхности подложки, выделяя пленкообразующие компоненты, причем подложка является более нагретым телом, чем окружающая среда. При выращивании тонких слоев методом ГФЭ МОС контроль параметров пленок непосредственно в процессе синтеза неприменим, так как обычно используют достаточно агрессивные газовые среды. Процесс проводят в проточном вакуумном реакторе при давлении паров компонентов 0,1–10 мм рт. ст. Снижение давления и увеличение скорости газового потока позволяют получать однородные слои с контролем толщины до нескольких ангстрем. Исходными компонентами обычно служат метильные, этильные и изобутильные соединения металлов, а также гидриды некоторых элементов.

В качестве примера можно привести химическиереакции, протекающие у поверхности при получении
тонких слоев полупроводников $\text{А$^\text{III}$В$^\text{V}$}$:

$$
\text{Ga(CH$_3$)$_3$ + AsH$_3$ $\rightarrow$ GaAs + 3CH$_4$},
$$

$$
\text{In(C$_2$H$_5$)$_3$ + PH$_3$ $\rightarrow$ InP + 3C$_2$H$_6$}.
$$

Смешивая исходные компоненты в газовой фазе в определенных пропорциях синтезируют эпитаксиальные пленки твердых растворов:

$$
(1-x)\text{Ga(CH$_3$)$_3$} + x\text{Al(CH$_3$)$_3$} + \text{AsH$_3$} \rightarrow \text{Ga$_{1-x}$Al$_x$As} + \text{3CH$_4$}.
$$

Добавляя в газовую фазу небольшие количества легколетучей примеси получают легированные эпитаксиальные слои.

МОС-гидридная технология, успешно конкурируя с МЛЭ своей простотой и производительностью, обладаенесколько худшей воспроизводимостью и дает более расплывчатые профили изменения состава и легирования. Это обусловлено использованием как сравнительно высоких температур подложки, так и длительным временем синтеза.

### Химическая сборка поверхностных наноструктур

Метод химической сборки (ХС) и его разновидности – метод молекулярного наслаивания (МН) и атомно-слоевая эпитаксия (АСЭ) – явились результатом поиска принципиально новых методов синтеза ультратонких слоев. Они основаны на образовании поверхностных химических соединений при хемосорбции компонентов из газовой фазы и являются новыми так называемыми циклично-дискретными процессами. Благодаря возможности реализовать практически монослойную
хемосорбцию компонентов формирование кристаллических структур происходит по слоевому механизму, то есть без образования трехмерных зародышей. Это позволяет получать сплошные пленки при толщинах в несколько монослоев вещества (менее 1 нм). Используют этот метод в основном для синтеза наноструктур бинарных соединений.

В методе молекулярного наслаивания предусмотрено наличие на поверхности матрицы определенных функциональных групп, которые способны реагировать с низкомолекулярным реагентом с образованием устойчивого соединения. Последнее также должно обладать активными группами, способными к дальнейшим взаимодействиям. Таким образом, благодаря серии последовательных химических реакций осуществляется наращивание слоев структурных единиц заданного состава. Основная особенность химической сборки заключается в том, что процесс формирования слоя контролируется не термодинамикой фазовых переходов, а термодинамикой макрореагентов химических реакций и проведение реакций в неравновесных условиях приводит к устойчивости
синтезируемой структуры. Неравновесность процесса легко достигается значительным избытком реагента и быстротой удаления газообразных продуктов реакции. Данный метод позволяет синтезировать нано- и микроструктуры на поверхности твердых веществ путем многократного чередования химических реакций по заданной программе. Толщина образующегося слоя определяется не временем синтеза или интенсивностью потока вещества, а количеством повторяющихся реакционных циклов. Благодаря малой энергии активации поверхностных реакций ХС обычно проводят при достаточно низких температурах (25–400°С), что значительно снижает влияние диффузионных процессов и позволяет получать многослойные структуры с резкими границами. Для предотвращения физической конденсации исходных реагентов $А$ и $В$ температура подложки $Т_S$ должна удовлетворять соотношению $Т_А, Т_В < Т_S < Т_{AB}$, в котором $Т_А, Т_В, Т_{AB}$ – критические температуры конденсации компонентов А, В и соединения АВ соответственно.

Как уже отмечалось, для осуществления процесса химической сборки по методу молекулярного наслаивания необходимо активировать поверхность функциональными группами, компоненты которых являются составляющими синтезируемого слоя. Так, группы –ОН используются для получения
оксидных слоев, –SH — сульфидных, –NH — нитридных и т.д. Они получаются обработкой поверхности соответствующими реагентами – парами $\text{Н$_2$О}$, $\text{Н$_2$S}$, $\text{NН$_3$}$. Полупроводниковые матрицы соединений $\text{А$^\text{III}$В$^\text{V}$}$ предварительно обрабатывают парами галогенводорода, что позволяет удалить оксидный слой и активировать поверхность галогенидными группировками. Процесс молекулярного проводят либо в проточном реакторе при атмосферном давлении, либо в системе с пониженным давлением.

В качестве примеров можно привести реакции для синтеза слоев.

### Химическая сборка поверхности твердых тел методом молекулярного наслаивания

Основная идея метода молекулярного наслаивания (МН), используемого для прецезионного синтеза твердых тел регулярного строения, состоит в последовательном наращивании монослоев структурных единиц заданного химического состава на поверхности твердофазной матрицы. Для осуществления указанного процесса необходимо соблюдение принципов, основными из которых являются следующие.
1. Синтез основывается на протекании необратимых в условиях эксперимента химических реакций между функциональными группами на поверхности твердого тела и молекулами подводимого извне реагента. При этом используемые реагенты и продукты реакции не должны вступать в химические взаимодействия между собой.
2. Для постепенного наращивания слоя нового вещества необходимо проводить многократную и попеременную (в заданной последовательности) обработку последнего парами соответствующих соединений. При этом каждый вновь образующийся
монослой новых функциональных групп должен содержать активные атомы или группы атомов, способные реагировать с новой порцией того же или иного реагента.
3. Для осуществления реакций в процессе МН необходимо некоторое структурное соответствие между поверхностью исходной твердофазной матрицы и данным соединением. Но главное – это наличие на поверхности как исходной, так и образующейся в ходе синтеза достаточного количества функциональных групп (ФГ) и с таким взаимным расположением, которое обусловливает возможность возникновения поперечных связей между присоединившимися атомами для образования трехмерной решетки синтезируемого твердого вещества.

На рисунке представлена схема процесса химической сборки поверхности твердого тела методом МН, иллюстрирующая его синтетические возможности исходя из изложенных принципов. Анализ представленных на рисунке схем и полученные в на-

Рассмотрим некоторые экспериментальные данные на примере синтеза методом МН оксидных структур на поверхности диоксида кремния.

Следующая операция – продувка сухим азотом для удаления избытка $\text{SiCl$_4$}$ и $\text{HCl}$. Если использовать хлориды других элементов, то можно получить структуры, содержащие атомы, отличные от имеющихся в матрице.

Число связей с поверхностью, концентрация привитых ФГ в монослое зависят от количества и расположения ОН-групп на исходной поверхности, химической природы реагента и будут уменьшаться с увеличением температуры. Отметим важнейшую закономерность метода МН: после замещения протона в составе всех доступных гидроксилов, сколько бы времени ни подавать тот же реагент в камеру, не удается получить (при соблюдении принципов МН) более одного монослоя новых ФГ. Существенный момент – проведение процесса в условиях максимального удаления от равновесия. ## Поверхностные фазы и поверхностная диффузия

Особое место среди последних достижений технологии занимает формирование структур с *поверхностными фазами* и с *дельта-легированными слоями*. Под *поверхностной фазой* следует понимать тонкий двумерный слой на поверхности монокристалла, находящийся в термодинамическом равновесии с его объемом, а *дельта-легирование* — это создание структуры, в которой
легирующая примесь находится в сверхтонком слое. В идеале толщина дельта-слоя совпадает с толщиной одного атомного слоя. Качественно новый шаг в развитии дельта-легирования связывают с упорядоченным  дельта-легированием, т. е. с формированием двумерных упорядоченных структур атомов легирующей примеси внутри дельта-слоя (при обычном, традиционном дельта-легировании атомы примеси распределены внутри дельта-слоя случайным образом).

Для решения задачи упорядоченного дельта-легирования предполагается использовать упорядоченные поверхностные фазы, которые формируются при осаждении субмонослойных пленок адсорбатов в условиях, близких к равновесным. Двухмерная структура атомов адсорбата (легирующей примеси) в поверхностной фазе является своего рода "заготовкой" для упорядоченного дельта-легирования, и в случае если при последующем наращивании покрывающего слоя эта структура будет сохранена на "захороненной" границе раздела, то поставленная задача будет решена. Такие структуры называются встроенными поверхностными фазами. Кроме улучшенных характеристик по сравнению с традиционными дельта-легированными слоями, встроенные поверхностные фазы могут иметь и особые свойства, так как известно, что сами поверхностные фазы часто обладают электронной структурой и физическими свойствами, отличными от объемных.

Наибольший интерес для полупроводниковой электроники представляет исследование поверхностных фаз на основе Si. При формировании поверхностных фаз, как и вообще при создании многослойных интегральных схем для полупроводниковых приборов, важную роль играет поверхностная диффузия. ## Максены

Экспериментальное получение графена механическим расслоением подтолкнуло исследователей к поиску новых (и развитию старых) методов получения двумерных материалов, таких как нитрид бора BN, дисульфиды молибдена MoS2 и вольфрама WS2, оксиды и гидроксиды переходных металлов. Поиски альтернатив графена связаны в первую очередь с тем, что материалы, состоящие из нескольких химических элементов, предоставляют большую вариативность в изменении их свойств, как за счет выбора компонентов, так и путем легирования, создания дефектов или функционализации различными химическими группами.

Рассматриваемое семейство материалов ‒ максены (англ. MXenes) — представляет собой слоистые карбиды, нитриды и карбонитриды, получаемые химическим расслоением материала-родителя состава $M_{n+1}AX_n$ (MAX-фазы), где $M$ ‒ переходный $d$-металл (Ti, V, Cr, Nb, Ta, Hf, Zr), $A$ ‒ элементы II–VI основных групп (Cd, Al, Ga, In, Tl, Si, Ge, Sn, Pb, P, As, S), $X$ ‒ углерод или азот (иногда используется терминология MAC-фаз для карбидов и $MAN$-фаз для нитри дов), число $n = 1, 2, 3...$. $MAX$-фазы соответствуют гексагональной пространственной группе $P63/mmc$ с двумя формульными единицами на элементарную ячейку, имеют слоистое строение и проявляют заметную анизотропию свойств.

Удаление $A$-слоя (рис. ) в таких MAX-материалах позволяет получить слабо связанные друг с другом двухмерные $MX$-слои, которые и называются максенами. Так как многие свойства максенов определяются свойства родительского материала, без получения которого невозможно рассматривать сами максены, начнем рассмотрение с MAX-материалов.

![](Figs/From_MAX_to_MXenes_structure.jpg)
*Рис. Схема преобразования MAX-фазы в MX-фазу (максен)*

В настоящий момент известно о существовании более 70 MAX-фаз. В зависимости от $n$ выделяют сооэтветствуют группы 211 ($n = 1$), 312 ($n = 2$), 413 ($n = 3$). Существуют также теоретические предсказания и единичные экспериментальные примеры существования MAX-фаз с большими $n$: 514 ($n = 4$), 615 ($n = 5$), и 716 ($n = 6$).

Структура MAX-материалов независимо от $n$ представляет плотно упакованные $M_{n+1}X_n$-слои, перемежающиеся слабо связанными с ними и потому химически более активными $A$-слоями, то есть материал представляет периодическую вдоль оси z структуру вида

$$.../[M_{n+1}X_n]/A/[M_{n+1}X_n]/A/...$$

При погружении образца MAX-фазы в плавиковую кислоту происходит
избирательное удаление слоя атомов элемента $A$, при этом соседние с ним слои травлению не подвергаются. Расслаивание $MX$-слоев происходит в результате, того, что при удалении $A$-слоя исчезает металлическая связь $M-A$.

Малая энергия связи $M-A$ проявляется не только при химическом травлении. При нагреве MAX-фазы происходит декомпозиция материала:

$$M_{n+1}AX_n \text{(тв.)} \rightarrow M_{n+1}X_n \text{(тв.)} + A \text{(г. или ж.)}$$

Температура разложения варьируется в широком диапазоне в зависимости
от природы MAX-фазы: от 850°C для $\text{Cr$_2$GaN}$ до 2300°C для $\text{Ti$_3$SiC$_2$}$.

Интерес к MAX-фазам самим по себе обусловлен тем, что их свойства являются некоторой комбинацией свойств керамических и металлических материалов — сочетая коррозионную стойкость и устойчивость к окислению, MAX-фазы одновременно проявляют высокие тепло- и электропроводность. Эти необычные свойства обусловлены смешанной ковалентно-металлической природой связи $M-X$ вместе со вместе со
слабой металлической связью $M-A$.

Так как формирующие материал MX-блоки являются карбидами и нитридами, в которых могут проявляться существенные отклонения от стехиометрии, MAX-фаза также может быть нестехиометричной. Наиболее распространенным типом дефектов являются вакансии в подрешетке $X$ соответствующие фазы описываются как $M_{n+1}AX_{n-\delta}$, где $\delta = 0...0.2$.

Важно отметить, что MAX-фазы образуют и твердые растворы, среди кото-
рых наиболее устойчивы $\text{(M1,M2)$_2$AlC}$, где $\text{M1}$ и $\text{M2 = Ti, V, Cr}$. Известно также о существовании множества твердых растворов вида $\text{Ti$_3$(Si,Al,Ge,Sn)C$_2$}$, то есть с
замещением атомов в подрешетке $A$. Среди твердых растворов с замещением в подрешетке $X$ наиболее важными являются карбонитриды, т. е. MAX-фазы вида $\text{M$_{n+1}$A(C,N)$_n$}$. Свойства твердых растворов существенно отличаются — так, $\text{Ti$_2$AlC$_{0.5}$N$_{0.5}$}$ является более твердым материалом, чем чистый нитрид $\text{Ti$_2$AlN}$ или карбид $\text{Ti$_2$AlC}$.

Основными способами получения тонкопленочных MAX-фаз являются:
* Физическое осаждение из паровой фазы (PVD): при температуре
подложки (обычно MgO (111)) 800−1200 C (в некоторых случаях — для $\text{Cr$_2$AlC}$ и $\text{V$_2$GeC}$ может быть снижена до 450 C) термическое, реактивное или магнетронное распыление одной или нескольких мишеней. Температура подложки может быть снижена в методах дугового осаждения и лазерного распыления мишени.
* Химическое осаждение из паровой фазы (CVD), например, в резуль-
тате разложения газовой смеси хлоридов (например, $\text{TiCl$_4$}$, $\text{SiCl$_4$}$, $\text{CCl$_4$}$ для получения $\text{Ti$_3$SiC$_2$}$ и водорода. Однако недостатком CVD-методики обычно является образование дополнительных не-MAX фаз, например, карбидов кремния и титана. Метод, также как и процесс физического осаждения требует высоких температур (1000 − 1300 C).
* Твердотельные реакции, проходящие, например, при нагреве контакта Ti и SiC (такие реакции известны при изготовлении электронных устройств на основе широкозонных материалов SiC и GaN с контактами из титана, когда на интерфейсе образуются соответственно $\text{Ti$_3$SiC$_ 2$}$ или $\text{Ti$_2$GaN}$).

Стоит заметить, что температуры синтеза тонких пленок существенно ниже, чем необходимые для роста объемных кристаллов. При этом, чем больше размер элементарной ячейки, тем выше температура, необходимая для формирования материала, так как требуется большая длина диффузии и тепловая энергия для ее активации. Таким образом, 211-фазы формируются при меньших температурах, чем структуры 312 и 413. Видимо, это служит одной из причин, почему низших MAX-фаз (то есть материалов с малыми $n$) известно существенно больше.

Наиболее доступной основой для синтеза максенов является $\text{Ti$_2$AlC}$. При травлении такой MAX-фазы получается одна из наиболее легких MX-фаз максенов ‒ $\text{Ti$_2$C}$. Английское название MXene было предложено по MX-фазе и аналогии с графеном (MX + graphene = MXene).

Распространенным подходом для получения максенов является погружение материала MAX-фазы в плавиковую (HF) кислоту. Рассмотрим соответствующие химические превращения на примере MAX-фазы $\text{Ti$_3$AlC$_2$}$:

$$\text{2Ti$_3$AlC$_2$ + 6HF = 2Ti$_3$C$_2$ + 2AlF$_3$ + 3H$_2$},$$

$$\text{Ti$_3$C$_2$ + 2H$_2$O = Ti$_3$C$_2$(OH)$_2$ + H$_2$},$$

$$\text{Ti$_3$C$_2$ + 2HF = Ti$_3$C$_2$F$_2$ + H$_2$}.$$

Первая реакция является освной и сопвровождается второей и/или третьей реакциями. Такой подход является упрощенным, так как две последнии реакции описываются через радикалы $\text{-OH}$, $\text{-F}$, в то время как при травлении радикалы могут участвовать совместно. В расслоенном $\text{Ti$_3$C$_2$}$ на каждую формульную единицу приходится два атома титана, открытых для функционализации.

![](Figs/SEM_of_MXene.png)
*Рис. Сканирующая электронная микроскопия образца максена, полученного травлением в $\text{HF}$ MAX-фазы $\text{Ti$_3$AlC$_2$}$*

В подавляющем большинстве работ по синтезу максенов в качестве травителя используется плавиковая кислота, однако для промышленного применения желательно использовать менее опасный травитель, например, гидрофторид аммония $\text{NH$_4$HF$_2$}$:

$$\text{2Ti$_3$AlC$_2$ + 6NH$_4$HF$_2$ = 2Ti$_3$C$_2$ + 2(NH$_4$)$_3$AlF$_6$ + 3H$_2$,}$$

$$\text{Ti$_3$C$_2$ + $a$NH$_4$HF$_2$ + $b$H$_2$O = (NH$_3$)$_c$(NH$_4$)$_d$Ti$_3$C$_2$(OH)$_x$F$_y$}.$$

Первая реакция соответствует травлению, вторая — интеркалированию. За счет медленного протекания реакции можно получать упорядоченные бездефектные слои.

Еще одним более мягким методом травления является нагрев максенов в растворах LiF и HCl. Этот метод позволяет получать чернила и пасты, создавать гибкую электронику на основе максенов. При этом получающиеся листы имеют большие латеральные размеры и не содержат дефектов, в отличие от обычного метода с использованием HF.

После применения того или иного способа химического расслаивания материал промывается деионизованной водой или растворов диметилсульфоксида, и проходит ультразвуковую обработку. Более эффективное расслаивание на отдельные слои достигается за счет интеркаляции диметилсульфоксида. Далее максенам можно придать необходимую форму при помощи холодного прессования.

Важно отметить, что структуры максенов получаются пока только метода-
ми травления, поэтому всегда функционализированы группами −O−, −OH и/или −F. С учетом этих групп получаемые травлением максены записывают как $M_{n+1}X_nT_x$, где $T$ — соответствующая группа или несколько групп.

### Свойства и применение максенов

Распределение плотности состояний максенов схоже с нестехиометрическими карбидами переходных металлов, составленных из глубоко лежащих $2s$-уровней углерода, гибридизированных $2p$-уровней и наиболее высоко лежащих $d$-уровней переходного металла, доминирующих вблизи $E_F$. Таким образом, электронные свойства MX-фазы в существенной мере определяются переходным металлом. Это более справедливо для «низших» максенов с $n = 1$, для максенов c $n = 2$ и $n = 3$ гибридизированные $2p$-электроны углерода оказывают большее
влияние. 

Функционализированные максены, получаемые на практике в результате
травления, представляют проводники или узкозонные полупроводники, чаще всего (среди исследованных случаев) непрямозонные. Металлические свойства максенов ослабляются по мере увеличения $n$, при этом нитриды проявляются более явные металлические свойства, чем карбиды.

Максены представляют интерес и как активные материалы ионисторов.
Показано, что ионистор на основе $\text{Ti$_3$C$_2$T$_x$}$ может обладать удельной емкостью до 300–500 Ф/см, как в щелочных, так и в кислотных электролитах при сохранении электрических свойств в течение $10^4$ циклов.

За счет развитой поверхности максены могут.

Из структуры и способа получения максенов напрашивается концепция супермаксенов ‒ атомарных сэндвичей, в которых максены различной природы, полученные методами top-down, являются строительными блоками для bottom-up-упорядочения в многослойные структуры. Однако получение таких структур пока затруднено тем, что для разделения слоев максенов
