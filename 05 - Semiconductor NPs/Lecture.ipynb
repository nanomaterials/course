# Конспект лекции (черновик)
## Квантовые точки

Основным параметром полупроводникового материала является значение ширины запрещенной зоны $E_g$. Способность изменения энергетических зазора определяет возможности изменения электронных и оптических свойств материала. В течение второй половины ХХ в. необходимого энергетического зазора добивались усложнением компонентногосостава полупроводниковых материалов: от бинарных соединений групп A$^\textrm{III}$B$^\textrm{V}$, A$^\textrm{II}$B$^\textrm{VI}$, A$^\textrm{IV}$B$^\textrm{VI}$ и др. к трех- и четырехкомпонентным твердым растворам.

На смену этой парадигме пришли низкоразмерные материалы, в которых изменение энергетического зазора осуществляется за счет пространственного ограничения материала в одном, двух или одновременно трех пространственных направлениях — соответственно 2D- (квантовые ямы), 1D- (квантовые нити) и 0D-материалы (квантовые точки).

Аналогично тому, как волны различной физической природы, возбуждаемые в конечном объеме, имеют строго определенную частоту, ограничение движения носителей заряда приводит к тому, что их энергии имеют дискретные значения — уровни разрешенных состояний. При критическом уменьшении размеров полупроводника в одном пространственном направлении непрерывный спектр плотности электронных состояний становится ступенчатым. Лишаясь второй степени свободы, полупроводник проявляет пикообразный спектр плотности состояний. Наконец, материал, ограниченный в трех пространственных направлениях, в силу сходства его спектра и спектра плотности атома, представляющих набор $\delta$-функций, называют искусственным атомом или квантовой точкой.

Термин «квантовая точка» был введен в 1988 г. Ридом при рассмотрении физических свойств структур, которые могут быть получены при литографии высокого разрешения. Хотя с позиций геометрии термин «квантовая точка» не является приемлемым, так как частица имеет некий размер и трехмерную структуру, в физике, где распространено модельное понятие «материальная точка», он допустим, и одновременно отражает очень малые размеры частицы и квантовую природу изменения ее энергетического спектра.

Таким образом, квантовая точка (КТ) — частица полупроводникового материала, обладающая зависящим от размера энергетическим спектром. Типичные размеры КТ составляют 1–20 нм, однако эти пределы существенно зависят от электронных свойств материала и структуры частицы. Наиболее важными с практической точки зрения свойствами КТ являются поглощение света в широком диапазоне длин волн при узкой линии эмиссии, а также зависимость длины волны флуоресценции от размера частиц.

## Квантово-размерные эффекты

При электронном возбуждении (например, при воздействии внешнего излучения), в полупроводнике образуются слабосвязанные электронно-дырочные пары —экситоны Ванье-Мотта. Область делокализации экситона, определяемая боровским радиусом экситона $r_B$.

$$
r_B = \frac{\hbar^2\varepsilon}{me^2},
$$

где $\hbar$ — редуцированная постоянная Планка, $\varepsilon$ — диэлектрическая проницаемость материала, $e$ — элементарный заряд, $m$ — приведенная масса экситона, выражаемая через эффективные массы электрона и дырки $m_e$ и $m_h$: $1/m = 1/m_e + 1/m_h$.

В силу того, что экситон является водородоподобной квазичастицей, модель описания энергетического строения экситона во многом сходна с предложенной Нильсом Бором моделью атома водорода. Поэтому удобной масштабной единицей области делокализации экситона является *боровский радиуc* — радиус ближайшей к ядру орбиты электрона в атоме водорода:

$$
a_0 = \frac{\hbar^2}{m_0e^2} = 0,0529 \text{ нм},
$$

где $m_0$ --- масса электрона. Как видно при подстановке $a_0$ в выражение для боровского радиуса экситона использование такой масштабной единицы облегчает сравнение $r_B$ для различных материалов:

$$
r_B = \frac{\varepsilon m_0}{m}a_0,
$$

Приведенная масса экситона $m$ для большинства полупроводников лежит в диапазоне десятых и сотых долей $m_0$, в то время как диэлектрическая проницаемость всегда больше единицы, поэтому для всех полупроводников выполняется условие $r_B \gg a_0$. Если радиус частицы полупроводника $r$ сопоставим с $r_B$, электронные свойства частицы значительно отличаются от свойств объемного материала. В этом случае говорят о *конфайнменте* — пространственном ограничении экситона и появлении квантово-размерных свойств. В то время как в массивных полупроводниках проявление экситонных состояний в спектрах поглощения ослабляется вследствие экранировки кулоновского взаимодействия $e^-$ и $h^+$ из-за высокой концентрации носителей заряда, в наночастицах число носителей заряда меньше и экситонные особенности ярко проявляются в спектре поглощения даже при комнатной температуре.

В случае, когда линейные размеры кристалла в трех измерениях оказываются меньше боровского радиуса экситона ($r \le r_B$), принято говорить о *сильном* пространственном ограничении. Энергетический спектр наночастицы в этом случае становится дискретным, а плотность состояний 0D-объекта представляет набор $\delta$-функций. На практике обычно исследуются спектры не отдельных частиц, а их ансамбля, например на подложке или в растворе — спектры поглощения и люминесценции оказываются уширенными вследствие наличия распределения частиц по размерам и отличной от абсолютного нуля температуры. Чем больше спектральных особенностей наблюдается в спектре поглощения, и чем лучше они разрешены, тем меньше дисперсия размеров частиц исследуемого образца.

Уменьшение размера частиц (при условии $r \le r_B$) вызывает эффект смещения экситонной полосы спектра поглощения (и всего спектра в целом) в высокочастотную область. Такое смещение в область меньших длин волн называют *синим сдвигом* (англ. blue shift). Аналогичный эффект проявляется в положении максимума интенсивности в спектре люминесценции квантовых точек.

В силу дискретного (или частично дискретного) характера энергетических уровней носителей заряда, использование таких понятий зонной теории как запрещенные и разрешенные зоны, не является корректным в отношении квантовых точек. Подчеркивая различие энергетического спектра полупроводниковых наночастиц и объемных кристаллов, говорят об *эффективной ширине запрещенной зоны* $E_g^*$. Поскольку наночастицы состоят из сравнительно небольшого количества атомов ($\sim 10^3$), по аналогии с молекулярными орбиталями, также рассматривают высшее занятое (HOMO) и низшее свободное (LUMO) состояния, разделенные энергетическим зазором.

Зависимость $E_g^*$ полупроводникового кристалла от его радиуса обычно описывается выражением, найденным Луи Брюсом из теории возмущения для малых кристаллов в приближении модели эффективной массы:

$$
E^*_g = E_g + \frac{h^2}{8mr^2}-\frac{1,8e^2}{4\pi\varepsilon\varepsilon_0r},
$$

где $E_g$ — запрещенная зона объемного материала, $h$ — постоянная Планка, $m$ — приведенная масса экситона, $r$ — радиус частицы. Последний член выражения описывает эффективное кулоновское взаимодействие электронно-дырочной пары.

Выражение позволяет из величины синего сдвига в спектре поглощения оценить средние размеры квантовых точек. Заметим, что выражение найдено Брюсом из представлений физики твердого тела, в частности понятия эффективной массы, и предполагает использование справочных значений эффективных масс носителей заряда объемных кристаллов, поэтому при строгом рассмотрении выражение не является вполне корректным для нанокристаллов, однако экспериментальные результаты в большинстве случаев достаточно хорошо согласуются с предложенным решением.

## Технология коллоидного синтеза квантовых точек

Путь синтеза коллоидных КТ лежит через создание коллоидных растворов с монодисперсными частицами посредством конденсации твердой фазы наночастиц из пересыщенного раствора. Рассмотрим наиболее общую
методику коллоидного синтеза полупроводниковых
нанокристаллов, называемую методом горячей инжекции.

Предварительно кратко опишем схему одного из вариантов установки, используемой для коллоидного синтеза КТ методом горячей инжекции (Рис.
3.5). Синтез проводится в трехгорлой колбе, поддержание температуры химической среды осуществляется при помощи колбонагревателя и системы термостатирования, считывающей значения с погруженной в раствор термопары. Однородность нагрева и распределения компонентов в
дисперсионной среде достигается перемешиванием при помощи магнитной мешалки. Описанная далее процедура впрыскивания прекурсора проводится через самоуплотняющуюся мембрану – септу. Для поддержания постоянного объема раствора используется система с обратным холодильником. Во избежание окисления реагентов синтез проводится в среде инертного газа.

В непрерывно перемешиваемую и нагретую до температуры нуклеации (T n = 200 ‒ 300 °C) дисперсионную среду, содержащую один из прекурсоров – источник ионов A, входящих в полупроводниковое соединение AB, впрыскивается состав, содержащий источник B. Например, в случае наночастиц селенида кадмия CdSe, в раствор прекурсора кадмия может впрыскиваться раствор прекурсора селена или наоборот. Прекурсоры образуют химически активные атомные или молекулярные фрагменты – мономеры: nA + nB = n(AB). Вследствие быстро протекающей реакции наблюдается локальное пересыщение раствора, в результате происходит нуклеация – образование зародышей твердой фазы.

На следующем этапе ядра растут, захватывая мономеры. Рост нанокристаллов из пересыщенного раствора происходит быстро, концентрация раствора практически мгновенно достигает уровня насыщения . После достижения насыщения может происходить созревание Оствальда – частицы, достигшие критического размера, растут за счет растворения более мелких частиц. Таким образом, система стремится к уменьшению свободной поверхностной энергии, наибольший вклад в которую вносят меньшие по размеру частицы. В каждый момент времени имеется некоторый критический размер, выше которого частицы растут, а ниже – растворяются. Этот критический размер не остается постоянным, а возрастает по мере уменьшения пересыщенности основным раствора. технологическим Процесс является параметром его диффузионным, поэтому регулирования является температура роста T g .

Размеры получаемых частиц технологически регулируются временем отбора частиц из реакционной среды относительно момента нуклеации – так в рамках одного технологического процесса можно получить серию образцов с различной длиной волны излучения за счет отбора частиц с различным диаметром и, соответственно, различным энергетическим спектром. Монодисперсность частиц обеспечивается одномоментной нуклеацией – частицы, зарождаясь в одно и то же время и испытывая одни условия роста, к моменту завершения процесса имеют один и тот же размер.

## Стабилизаторы 

Отдельного внимания заслуживает вопрос стабилизации – необходимо достичь не только заданного распределения размеров частиц, но и стабилизировать коллоидную систему, начиная с самых ранних этапов синтеза. Этого можно добиться за счет уменьшения энергии поверхностного натяжения на границе раздела фаз, что ведет к необходимости присутствия в системе кроме дисперсионной среды растворителя и дисперсной фазы полупроводниковых частиц еще одного компонента – молекул поверхностно- активного вещества. Адсорбируясь на границе раздела фаз, они создают барьер, препятствующий агрегации частиц, и пассивируют поверхностные энергетические состояния нанокристаллов, что снижает вероятность процессов безызлучательной рекомбинации и повышает квантовый выход люминесценции частиц. Лиганды влияют на скорость роста, морфологию, пути протекания реакции и распределение частиц по размеру.

Электронная структура пассивирующих поверхность частицы молекул лигандов влияет на электронные и оптические свойства системы. Наиболее яркий пример важности рассмотрения соотношения энергетических уровней неорганического ядра и покрывающих лигандов представляет система CdTe/тиол, где молекулы тиолов подавляют уход носителей заряда из полупроводникового ядра, вследствие чего структура обладает высокой квантовой эффективностью даже в отсутствии дополнительных неорганических оболочек. Сульфид и селенид кадмия таких особенностей с тиоловыми лигандами не проявляют. Выбор стабилизатора имеет решающее значение в коллоидном синтезе КТ. К молекулам стабилизатора предъявляется ряд важных требований – они должны:

1) предотвращать агрегацию частиц;
2) не препятствовать диффузионному росту;
3) адсорбируясь на поверхности частицы, образовывать прочную ковалентную связь с одной стороны, в то время как свободный конец молекулы должен иметь сильное сродство к диспергируемость КТ; 4) пассивировать поверхность КТ.


Проблема поиска подходящих стабилизирующих веществ была основным
препятствием на пути создания монодисперсных КТ до 1993 г., пока в научной
группе Бавенди не было предложено использовать для синтеза КТ
халькогенидов кадмия в качестве стабилизаторов триоктилфосфин TOP и его
оксид TOPO (Табл. 3.1). Использование этих стабилизаторов позволило
получать КТ CdS, CdSe, CdTe с узким распределением по размерам и высоким
квантовым выходом флуоресценции.

Преимущества использования TOP и TOPO состоят в возможности
проведения синтеза при сравнительно высоких температурах (до 350 °С), что
способствует формированию бездефектной кристаллической решетки. Прочная
связь молекул TOPO с поверхностью полупроводника позволяет циклически
проводить процедуры осаждения и редиспергирования без существенных
изменений свойств частиц. Лиганды также эффективно блокируют электронные
поверхностные состояния, обеспечивая яркую узкую линию люминесценции.

Метод горячей инжекции обладает рядом преимуществ перед другими методами синтеза наночастиц:
1) позволяет получать монодисперсные наночастицы с отклонением
среднего размера не более 5 %;
2) частицы могут быть получены в виде порошка и диспергированы в
различных матрицах после самого синтеза либо упорядочены на твердой
подложке;
3) контролируя рост частиц, можно при необходимости в одном технологическом процессе получить частицы различных размеров,
4) температуры синтеза не превышают 400 °С.

При подборе экспериментальных условий (комбинация прекурсоров, ПАВ, растворитель, температура и т.д.) можно синтезировать КТ различного II VI IV VI состава, в том числе нанокристаллы бинарных соединений A B , A B , III V A B и их твердых растворов. Технология позволяет получать наночастицы не только с разными составом и размерами, но и различной формы

Изоляция КТ и размерно-селективное осаждение. Непосредственно после того, как получен коллоидный раствор, содержащий частицы нужных размеров, его необходимо очистить от побочных продуктов и непрореагировавших компонентов, чтобы остановить рост частиц, который медленно может продолжаться при комнатной температуре и получить коллоидный раствор, содержащий только частицы и растворитель. Решение задачи состоит во введении сильно полярного вещества – осадителя, чаще всего безводного метанола. Частицы из-за покрывающих их молекул ПАВ, обращенных гидрофобной частью в раствор, неустойчивы в среде полярного растворителя и флокулируют (выпадают в осадок), в то время как легкие молекулы и ионы непрореагировавших веществ остаются в растворе. Флокуляцию ускоряют за счет центрифугирования раствора.

Аналогичный подход может быть применен для разделения частиц по размеру. При введении меньшего количества осадителя в первую очередь флокулируют самые крупные частицы, поскольку между ними сильнее взаимодействие Ван-дер-ваальса. Для этого в очищенный раствор КТ медленно добавляется осадитель до появления опалесценции (рассеяния света), означающей начало образования флокулята. Отделяя центрифугированием осадок (более крупные частицы) и раствор (более мелкие) и заново диспергируя осадок, можно получить два коллоидных раствора с меньшим разбросом размеров частиц, чем исходно. Повторение этой процедуры позволяет получить растворы частиц со все более узкой дисперсией размеров.


## Особенности легирования наночастиц

Легирование полупроводников лежит в основе микроэлектроники. Обсудим, что представляет собой легирование наночастиц. Нанокристалл размером 4 нм содержит $\sim$1000 атомов. Добавление одного единственного атома примеси приводит к концентрации\footnote{Концентрации атомов в кристалле $\sim10^{22}$ см$^{-3}$} примеси $\sim10^{19}$ см$^{-3}$ . В объемном полупроводнике такое значение уже соответствовало бы сильно легированному, вырожденному полупроводнику. Однако необходимо учитывать, что с уменьшением зазоров происходит увеличение энергетического зазора и сам вопрос не является тривиальным

# Стоксов сдвиг

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
ABS = pd.read_csv("Spectra/CsPbBr3-ABS.csv", delimiter=',')
PL = pd.read_csv("Spectra/CsPbBr3-PL.csv", delimiter='\t', header=None,
                names = ['Wavelength', 'I'])
PL['I'] = PL['I'] - PL['I'].min()
PL['I'] = PL['I']/PL['I'].max()
plt.plot(ABS['Д. в., нм'], ABS['A'], label='Поглощение')
plt.plot(PL['Wavelength'], PL['I'], label='Люминесценция')
plt.xlabel('$\lambda$, нм')
plt.legend()
plt.show()
```

Хотя спектральные зависимости часто приводят в виде зависимостей от длины волны, физическое рассмотрение в случае полупроводников практически всегда удобнее вести в терминах энергий — в данном случае энергий фотонов:

$$
E = h\nu = \dfrac{hc}{\lambda}.
$$

В этом случае спектры "перевернутся" слева направо и слегка растянутся, но зато мы получим линейный масштаб относительно шкалы энергиий: 

```python
h = 6.62E-34 # постоянная Планка
c = 3E+8     # скорость света
q = 1.6E-19  # заряд электрона (для перевода из Дж в эВ)

E_ABS = h*c/q*10**9/ABS['Д. в., нм']
E_PL = h*c/q*10**9/PL['Wavelength']
plt.figure(figsize=(5, 5))
plt.plot(E_ABS, ABS['A']*1.35, label='Поглощение')
plt.plot(E_PL, PL['I'], label='Люминесценция')
plt.xlabel('$E$, эВ')
plt.legend()
plt.savefig('CsPbBr3-170C.png')
plt.show()
```

Первая производная...

```python
diff_ABS = -np.gradient(ABS['A'])
diff_PL = -np.gradient(PL['I'])
plt.plot(E_ABS, diff_ABS)
plt.plot(E_PL, diff_PL)
plt.xlabel('$E$, эВ')
plt.ylabel('$\dfrac{dA}{dE}$, $\dfrac{dI}{dE}$')
plt.show()
```

Взятие второй производной сводит операцию нахождения Стоксова сдвига к разности между энергетическими положениями первых двух максимумов вторых производных.

```python
diff2_ABS = -np.gradient(np.gradient(ABS['A']))
diff2_PL = -np.gradient(np.gradient(PL['I']))

plt.plot(E_ABS, diff2_ABS)
plt.plot(E_PL, diff2_PL)

# расскомментируйте две следующие строки,
# если хотите сравнить положение исходных кривых и вторых производных
# plt.plot(E_ABS, ABS['A']/150, label='Поглощение')
# plt.plot(E_PL, PL['I']/100, label='Люминесценция')

plt.xlabel('$E$, эВ')
plt.ylabel('$\dfrac{d^2A}{dE^2}$, $\dfrac{d^2I}{dE^2}$')
plt.show()



ABS_max = E_ABS[np.argmax(diff2_ABS)]
PL_max = E_PL[np.argmax(diff2_PL)]

print("""
Положение первого максимума в спектре поглощения - {:.3f} эВ.
Положение максимума фотолюминесценции - {:.3f} эВ.
Таким образом, Стоксов сдвиг составляет {:.0f} мэВ.
""".format(ABS_max, PL_max, (ABS_max-PL_max)*1000))
```

```python
import os
os.chdir('../Data')


import numpy as np
import pandas as pd    # библиотека для работы с табличными данными
import matplotlib.pyplot as plt

materials = pd.read_csv("semiconductors.csv")
materials.head(15)                              # 10 первых записей таблицы 
```

К сожалению, далеко не для всех полупроводников материалов экспериментально исследованы их физические свойства (пропущенные табличные значение в Pandas соответствуют типу NaN - Not a Number).
```python
materials.columns
```

```python
materials = materials[(materials['Тип'] == 'D') &
                      (materials['me*'].notnull()) &
                      (materials['mh*'].notnull()) &
                      (materials['epsilon(infty)'].notnull())]
materials = materials[['Материал', 'Формула', 'Eg, эВ', 'me*', 'mh*', 'epsilon(infty)']]
```

```python
materials['aB'] = np.round(materials['epsilon(infty)']*(1/materials['me*'] + 1/materials['mh*'])*0.0529, 1)
materials
```

```python
%matplotlib inline
h = 6.62E-34 # постоянная Планка в Дж*с
m0 = 9.1E-31 # масса электрона в кг
q = 1.6E-19  # заряд электрона
eps0 = 8.85E-12

r = np.linspace(0.5, 100, 1000)*1E-9
plt.figure()

def QD_gap(Eg, me, mh, r, eps):
    exiton_energy = h**2/(8*m0*r**2)*(1/me + 1/mh)/q
    q_energy = -1.8*q**2/(4*np.pi*eps0*eps*r)/q
    return Eg + exiton_energy + q_energy

for row in materials.index:
    Material, Formula, Eg, me, mh, eps, a = materials.loc[row]
    plt.plot(r/1E-9, QD_gap(Eg, me, mh, r, eps), label=Formula)

plt.xlabel("$r$, нм")
plt.ylabel("$E^*_g$, эВ")
plt.ylim(0, 4)
plt.xlim(0, 10)
plt.legend()
plt.show()
```
