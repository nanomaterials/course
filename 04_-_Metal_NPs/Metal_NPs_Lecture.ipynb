## Оптические свойства металлических наночастиц

В электронных спектрах поглощения металлических наночастиц доминируют полосы поглощения поверхностных плазмонов. *Плазмон* — это коллективные колебания электронов, возбуждаемые переменным электромагнитным полем, относительно ионного остова с определенной резонансной частотой $\omega_\text{p}$, называемой плазмонной:

$$
\omega_\text{p} = \sqrt{\frac{ne^2}{m_e\varepsilon_\text{0}}},
$$

где $n$ — концентрация электронов, $m_e$ — эффективная масса электрона.

Если частота внешнего излучения больше $\omega_\text{p}$, электроны совершают колебания и на данных частотах металл прозрачен для внешнего излучения. При частотах внешнего излучения меньше $\omega_\text{p}$ электроны совершают колебания противоположно полю внешнего излучения, тем самым вызывая отражение на данных частотах. В случае наночастиц движение электронов пространственно ограничено по всем трем направлениям, что при взаимодействии металлической наночастицы с внешним излучением приводит к образованию локализованного плазмона. 

При условии $\lambda \gg r$ можно считать, что электромагнитное поле постоянно вблизи частицы. При воздействии внешнего поля на частицу электроны проводимости смещаются к границе частицы в направлении, противоположном направлению поля. В следствие того, что электроны пространственно ограничены, отрицательный заряд будет накапливаться на одной стороне частицы, а положительный — на другой, создавая таким образом электрический диполь, поле которого будет направлено противоположно внешнему полю. Данное поле будет вынуждать электроны вернуться в исходное положение, в результате чего будут происходит колебания электронов.

При колебаниях электроны рассеиваются на границах частицы и на ионах решетки, из-за чего теряют свою энергию. Чтобы восполнить эту энергию, электроны поглощают энергию внешнего поля. Плазмонный резонанс сопровождается резонансным поглощением, которое можно зафиксировать, измеряя спектр оптической плотности.

ЛПР описывается поперечным сечением экстинкции $C_\text{ext}$, которое для металлической частицы сферической формы имеет вид:

$$
C_\text{ext} = \frac{24\pi^2 r^3 \varepsilon_\text{m}^{3/2}}{\lambda} \frac{\varepsilon_\text{i}}{(\varepsilon_\text{r} + 2\varepsilon_\text{m})^2 + \varepsilon_\text{i}^2}.
$$

где $\varepsilon_\text{m}$ — относительная диэлектрическая проницаемость среды, $\varepsilon_\text{r}$ — действительная часть диэлектрической проницаемости металла, $\varepsilon_\text{i}$ — мнимая часть диэлектрической проницаемости металла.

Резонансное увеличение сечения экстинкции наблюдается при следующем условии:

$$
\varepsilon_\text{r} = -2\varepsilon_\text{m}.
$$

Чтобы в частице поддерживался заметный плазмонный резонанс, $\varepsilon_\text{i}$ должна быть минимальна. Зависимость действительной части диэлектрической проницаемости от частоты описывается выражением

$$
\varepsilon_\text{r} = 1 - \frac{\omega^2_\text{p}}{\omega^2}.
$$

Приравняв два последних выражения, получаем

$$
\omega_\text{max} = \frac{\omega_\text{p}}{\sqrt{1 + 2\varepsilon_\text{m}}}.
$$

Рассчитаем спектры компонент диэлектрической проницаемости серебра в диапазоне длин волн 200—1000 нм по данным о спектрах действительной и мнимной компонент показателя преломления.

```python
%pwd

os.getcwd()

import os
import csv

os.chdir('../Data/refractiveindex/')
```

```python
for entry in os.scandir():
    if entry.is_file() and 'Ag' in entry.name:
        Ag_path = entry.name
    if entry.is_file() and 'H2O' in entry.name:
        H2O_path = entry.name
```

```python
import numpy as np
import matplotlib.pyplot as plt

def wl_nk(path):
    '''Считывает файл с данными формата с сайта refractiveindex.info,
    возвращает таблицу значений вида длина волны - n - k.'''
    data = np.genfromtxt(path, dtype=float, delimiter=',', names=True) 
    wl_n = data[:data.size//2]
    wl_k = data[data.size//2+1:]
    wl_n = np.array([list(wln) for wln in wl_n])
    wl_k = np.array([list(wlk) for wlk in wl_k])
    if np.array_equal(wl_n[:,0], wl_k[:,0]):
        return np.hstack((wl_n, wl_k[:, 1:2]))
    

def wl_eps(path):
    '''Преобразует табличные спектры компонент показателя преломления
    к спектрам компонент диэлектрической проницаемости'''
    wlnk = wl_nk(path)
    wl, n, k = wlnk[:, 0], wlnk[:, 1], wlnk[:, 2]
    eps = (n + 1j*k)**2
    re_eps = np.real(eps)
    im_eps = np.imag(eps)
    return np.stack([wl, re_eps, im_eps], axis=1)

def show_eps_spectra(path):
    eps = wl_eps(path)
    plt.plot(eps[:, 0]*1000, eps[:, 1], label="$\epsilon'$")
    plt.plot(eps[:, 0]*1000, eps[:, 2], label="$\epsilon''$")
    plt.xlabel('$\lambda$, нм ')
    plt.ylabel("$\epsilon', \epsilon''$")
    plt.legend()


show_eps_spectra(Ag_path)
plt.xlim([200, 1000])
plt.ylim([-40, 5])
plt.show()
```

Как видно, действительная часть диэлектрической проницаемости серебра отрицательна в видимом диапазоне, что означает возможность выполнения условия резонанса, при этом мнимая часть диэлектрической проницаемости мала. В качестве среды рассмотрим дистилированную воду. Мнимая составляющая диэлектрической проницаемости в видимом диапазоне равна нулю, действительная составляющая имеет положительное значение менее 2.

```python
show_eps_spectra(H2O_path)
plt.xlim([200, 1000])
plt.show()
```

Из условия резонанса видно, что положение плазмонного резонанса определяется также диэлектрической проницаемостью окружающей среды. Диэлектрическая среда поляризуется под воздействием поля, в результате чего на границе металл/диэлектрик накапливается заряд, частично компенсирующий заряд на границах частицы, что приводит к уменьшению создаваемого ей поля. Чем больше диэлектрическая проницаемость среды, тем больше компенсация заряда и тем меньше поле, создаваемое частицей. Поэтому при увеличении диэлектрической проницаемости среды, окружающей металлическую частицу, происходит смещение максимума плазмонного резонанса в длинноволновую область.

Сравним влияние различных параметров на спектр экстинкции металлических наночастиц
* размер наночастиц
* материал наночастиц
* материал дисперсионной среды

```python
from scipy.interpolate import interp1d

def C_ext_spectra(r, metal, medium, label):
    '''Находит спектр экстинкции плазмонной наночастицы для переданных материалов.'''
    
    # считывание табличных спектров диэлектрической проницаемости
    for entry in os.scandir():
        if entry.is_file() and metal in entry.name:
            met_wl_eps = wl_eps(entry.name)
        if entry.is_file() and medium in entry.name:
            med_wl_eps = wl_eps(entry.name)
    met_wl, met_re_eps, met_im_eps = met_wl_eps[:, 0], met_wl_eps[:, 1], met_wl_eps[:, 2]
    med_wl, med_re_eps, med_im_eps = med_wl_eps[:, 0], med_wl_eps[:, 1], med_wl_eps[:, 2]
    
    # нахождение интерполяционных функций с кубическими сплайнами
    met_re_eps = interp1d(met_wl, met_re_eps, kind='cubic')
    met_im_eps = interp1d(met_wl, met_im_eps, kind='cubic')
    med_re_eps = interp1d(med_wl, med_re_eps, kind='cubic')
    
    # определение диапазона, на котором возможен адекватный расчет
    wl_min = max(np.min(met_wl), np.min(med_wl))   
    wl_max = min(np.max(met_wl), np.max(med_wl))
    wl = np.linspace(wl_min, wl_max, 100000)
    
    # расчетная формула
    C = 24*np.pi**2*r**3/wl/1e6  # данные в файлах приведены в мкм 
    C = C*med_re_eps(wl)**1.5*met_im_eps(wl)
    C = C/((met_re_eps(wl) + 2*med_re_eps(wl))**2 + met_im_eps(wl)**2)
    
    plt.xlabel('$\lambda$, нм')
    plt.ylabel("$C_{ext}'$")
    plt.xlim([200, 1000])
    plt.plot(wl*1000, C, label=label)
    return wl, C
```

```python
for metal in ['Ag', 'Au', 'Cu']:
    C_ext_spectra(10e-9, metal, 'H2O', metal)

plt.legend()
plt.show()
```

```python
for medium in ['H2O', 'SiO2', 'Al2O3', 'TiO2']:
    C_ext_spectra(10e-9, 'Ag', medium, medium)

plt.legend()
plt.show()
```

Плазмонный резонанс наблюдается у многих металлов — в особенности у металлов с наименьшим удельным сопротивлением — серебра, золота, меди и алюминия. Особый интерес представляет изучение ЛПР серебрянных наночастиц, так как серебро обладает наибольшими тепло- и электропроводностью среди данных материалов, а также обладает наиболее сильным резонансом. Наблюдение у частиц серебра сильного плазмонного резонанса объясняется двумя причинами. Во-первых, сила ПР может быть описана при помощи фактора качества

$$
Q = \dfrac{w}{2\varepsilon_\text{i}^2}\dfrac{d\varepsilon_\text{r}}{dw}.
$$

```python
def Q_spectra(metal):
    # считывание табличных спектров диэлектрической проницаемости
    for entry in os.scandir():
        if entry.is_file() and metal in entry.name:
            met_wl_eps = wl_eps(entry.name)
    met_wl, met_re_eps, met_im_eps = met_wl_eps[:, 0], met_wl_eps[:, 1], met_wl_eps[:, 2]
   
    omega = 2*np.pi*3e8/met_wl/1e-6
    D = np.diff(met_re_eps)/np.diff(omega)
    Q = omega[1:]/(2*met_im_eps[1:]**2)*D
    plt.xlabel('$\omega$')
    plt.ylabel("$Q$")
    plt.plot(omega[1:], Q, label=metal)

for metal in ['Ag', 'Au', 'Cu']:
    Q_spectra(metal)

plt.legend()
plt.show()
```

Серебро обладает наибольшим $Q$ среди данных металлов. Во-вторых, важную роль в тушении ПР играют межзонные переходы. В случае серебра энергии, соответствующие межзонным переходам больше энергий ПР, что приводит к минимальному тушению плазмонов. Для серебра начало межзонных переходов можно наблюдать в спектре экстинкции как небольшое увеличение поглощения, начиная примерно с 330 нм. У золотых частиц ПР наблюдается на длине волны 530 нм, что очень близко к краю области межзонных переходов. У частиц меди наблюдается самый слабый ПР среди перечисленных металлов, так как он перекрывается с областью межзонных переходов.

Как было упомянуто ранее, положение плазмонного резонанса определяется также размерами и геометрией частиц. Плазмонными свойствами могут обладать как сферические частицы, так и частицы, имеющие более сложную геометрию, такие как треугольные призмы, кубические частицы и т.:д., в том числе гибридные структуры — оболочки металла на поверхности сферической частицы другого материала. В зависимости от геометрии частиц могут проявляться дополнительные частоты ЛПР, а также смещение основной частоты. Так, в случае оболочечных структур, меняя их геометрию, можно получить ЛПР в широкой области видимого и инфракрасного излучения. При уменьшении симметрии частицы увеличивается число пиков ЛПР, что связано с тем, что при более низкой симметрии структуры у электронов появляется возможность сместиться более, чем в одном направлении. Наличие острых углов в геометрии частицы смещает ЛПР в длинноволновую область, так как увеличивается разделение заряда и уменьшается восстанавливающая сила колебаний. По аналогичной причине ЛПР смещается в длинноволновую область при увеличении размера сферических частиц. 

## Эффект усиления электромагнитного поля вблизи плазмонных наночастиц

При плазмонном резонансе металлической наночастицы происходит усиление электромагнитного поля в ближнем поле частицы по сравнению с полем падающей волны. Рассмотрим сферические частицы радиусом $r$, на которые падает поляризованный по оси $z$ свет с длиной волны $\lambda$. Решение для ближнего электромагнитного поля частицы может быть записано следующим образом

$$
E_\text{out} = E_0 \vec{z} - \left[\frac{\varepsilon - \varepsilon_\text{m}}{\varepsilon + 2\varepsilon_\text{m}}\right] r^3 E_0 \cdot \left[\frac{\vec{z}}{l^3} - \frac{3z}{l^5}(x \vec{x} + y \vec{y} + z \vec{z})\right],
$$

где $E_0$ — амплитуда падающего электрического поля, $\varepsilon$ — диэлектрическая проницаемость металлической наночастицы, $\vec{x}, \vec{y}, \vec{z}$ — единичные векторы, $l$ — радиус-вектор положения частицы в пространстве.

Так как интенсивность возбуждения флуорофора пропорциональна квадрату интенсивности электромагнитного поля, происходит усиление люминесценции, так как флуорофоры будут поглощать больше света. При этом максимальное усиление достигается, если энергия ЛПР равна энергии поглощения флуорофора. Кроме того металлические наночастицы могут изменять соотношение скоростей излучательных и безызлучательных переходов флуорофора, влияя на его квантовый выход и время жизни носителя заряда в возбужденном состоянии.

Однако при слишком малых расстояниях между плазмонной частицей и люминесцирующим объектом происходит ослабление люминесценции до исходного или даже меньшего уровня. Одна из причин уменьшения интенсивности фотолюминесценции --- носитель заряда из фотовозбужденной пары электрон-дырка в двухуровневой структуре может туннелировать и быть захвачен металлической наночастицей, что приводит к безызлучательной рекомбинации. Увеличения потенциального барьера между флуорофором и металлической наночастицей и соответствующего снижения вероятности туннелирования заряда добиваются изменением толщины разделяющего слоя (спейсера). Но при этом эффект усиления носит ближнепольный характер и быстро спадает при увеличения расстояния от плазмонной частицы. Таким образом, существует некоторая оптимальная толщина спейсера $h_\text{opt}$ (обычно порядка нескольких нм), при которой будет наблюдаться наибольшее усиление.

### Гигантское комбинационное рассеяние

Прежде чем говорить о гигантском комбинационном рассеянии света (ГКР), напомним о явлении традиционного [комбинационного рассеяния](https://ru.wikipedia.org/wiki/Комбинационное_рассеяние_света) (КР) света, называемого еще рамановским рассеянием. Явление КР света заключается в появлении в спектре света, рассеянного средой при ее возбуждении монохроматической накачкой, новых спектральных компонент, таких, что разности частот этих спектральных линий в спектре КР, называемых стоксовыми $\omega_{st, i}$ и антистоксовыми $\omega_{ast, i}$ и частоты возбуждающего света $\omega_L$ совпадают с частотами внутримолекулярных колебаний $\Omega_i$, набор которых, в свою очередь, является однозначной характеристикой молекул среды. При этом $\omega_{st, i} = \omega_L - \Omega_i$ и $\omega_{st, i} = \omega_L + \Omega_i$.

Интенсивность рассеянного света определяется соответствующими молекулярными восприимчивостями и числом молекул, попадающих в объем рассеяния (которое, в свою очередь, зависит от плотности вещества, то есть и от его агрегатного состояния). Типичное отношение интенсивности стоксовой компоненты к интенсивности подающего излучения в конденсированных средах (в жидкостях и твердых телах) составляет $\dfrac{I(\omega_{st})}{I(\omega_L)} = 10^{-6}$, что позволяет считать КР весьма слабым эффектом.

Малость вероятности процесса КР делает его наблюдение даже при исследовании конденсированной фазы достаточно сложной задачей, требующей современных лазерных источников и систем счета фотонов для регистрации рассеянного излучения. Кроме того, КР является фантастически эффективным инструментом исследования строения молекул, поскольку набор частот внутримолекулярных колебаний однозначно связан со структурой молекул и внутримолекулярным взаимодействием. Это заставляет думать о применении КР для исследования систем с малым числом молекул в объеме рассеяния, например в молекулярных пучках или моно- и субмонослоях молекул, адсорбированных
на поверхности. В последнем случае грубые оценки показывают, что число участвующих в процессе рассеяния молекул на шесть-семь порядков меньше, чем для
твердого тела или жидкости. Тогда для монослоя адсорбата $\dfrac{I(\omega_{st})}{I(\omega_L)} = 10^{-12}-10^{-13}$.

Однако идея исследования адсорбатов методом КР-спектроскопии была настолько привлекательной, что экспериментаторы пытались разными ухищрениями увеличить регистрируемый сигнал КР. В 1974 г. английский электрохимик М. Флейшман решил увеличить эффективное число молекул, участвующих в рассеянии от монослоя, увеличив площадь поверхности при сохранении площади, освещаемой излучением накачки. Для этого он прошероховатил поверхность серебра в электролите (в водном растворе KCl) методом анодного травления, а затем там же в электрохимической ячейке адсорбировал на разросшуюся поверхность серебра монослой молекул пиридина $\text{C$_5$H$_5$N}$. М. Флейшману легко удалось наблюдать спектр КР адсорбированных молекул, что он интерпретировал как следствие возрастания эффективной площади монослоя. Однако в конце 1970-х годов “вдруг” стало понятно, что что-то здесь не так: регистрируемая интенсивность КР по аккуратным оценкам возрастала в $10^6-10^7$ раз, в то время как площадь разрыхленной анодным травлением поверхности серебряного электрода увеличивалась только на порядок. Это говорило о том, что интерпретация М. Флейшмана его собственных экспериментов неверна (или неполна, по крайней мере) и за наблюдаемым усилением КР может стоять новый нетривиальный эффект, который и был назван позднее гигантским КР.

## Физические методы синтеза металлических наночастиц

- **Конденсация в инертном газе**. Металлическая фольграф или иной источник металла нагреваются в керамическом тигле, находящемся в камере, наполненной инертным газом (обычно аргоном) при давлении несколько торр. Металлический пар быстро охлаждается, теряя энергию в столкновениях с атоами аргона, при этом формируются наночастицы. Таким способом получаются, например, наночастицы Fe, Fe2O3.
- **Метод дугового разряда**. Для испарения материала используется дуга между металлическими электродами в присутствии инертного газа, в результате чего формируется плазма, при остывании которой образуются наночастицы металла. Так были получены наночастицы Ni.
- **Лазерный пиролиз**. Газообразный прекурсор смешивается с инертным газом и нагревается непрерывным или импульсным инфракрасным CO2-лазером. Энергия поглощается либо прекурсором, либо добавляемым в смесь инертным газом сенсибилизатором, например SF6. Сильный негрев газа приводит к диссоциации прекурсора и образованию атомов металла, которые охлаждаются при столкновении с атомами инертного газа и агрегируют, формируя наночастицы.
- **Струйный пиролиз**. Для получения наночастиц сформированные с помощью распылителя микрокапли раствора прекурсора инжектируются в горячую зону печи. Размер наночастиц зависит от температуры распылителя, относительного давления газов и температуры печи. Указанным способом были получены НЧ Cu.

Заметим, что добавление соответствующих источник неметаллов или использованием вместо инертного газа смеси инертного газа с газообразным прекурсом неметалла можно получить наночастицы полупроводника.

## Хмические методы синтеза металлических наночастиц
- **Восстановление ионов металла**. Один из наиболее распространенных подходов к синтезу металлических наночастиц это восстановление металлов из их солей с ограничением роста нанокристаллов с помощью поверхностно-активных веществ  или ионов.
    - *Восстановление боргидридом натрия*, известное с 1950-х гг., было одним из объектов исследования в [Манхэттенском проекте](https://ru.wikipedia.org/wiki/Манхэттенский_проект). Основная реакция состоит в гиролизи гедрида бора, сопровождаемом выделением водорода:
    BH$_4^-$ + 2H$_2$O $\rightarrow$ BO$_2^-$ + 4H$_2$.
    - *Восстановление цитратом натрия* наиболее популярно для получения гирозолей Au. Синтез наночастиц этим методом основан на том, что золотохлористая кислота (HAuCl$_4$) при нагреве смешивается с раствором цитрата натрия (Na$_3$C$_6$H$_5$O$_7$). Возникновение у раствора характерного цвета красного вина указывает на протекание реакции восстановления. Метод также используется для получения наночастиц Pt, Ag.
    - *Восстановление высокомолекулярными спиртами (полиолами)*, например, этиленгликолем используется для получения сравнительно крупных наночастиц с размерами от нескольких десятков нм до единиц микрометров. 
- **Фотохимический синтез**. Разложение светом металлических комплексов (фотолиз) или восстановление солей металлов за счет генерирования светом восстановителей, таких как сольватированные электроны (радиолиз). Так, например, при облучени ультрафиолетом HAuCl$_4$ в формамиде можно получить НЧ Au.
- **Электрохимический синтез**. Метод представляет усовершенствование классического процесса электрообработки и включает следующие стадии: окислительное разложение анода, миграцию металлическию частиц на катод, восстановление ионов до нульвалентного состояния, формирование частиц за счет нуклеации и роста, остановку роста с помощью улавливающих веществ и осаждение частиц. Этим методом были получены металлические НЧ разного размера (начиная от 1—5 нм): Pd, Ni, Co, Fe, Ti, Ag и Au. При использовании двух анодов из соответствующих материалов [можно получать](https://pubs.acs.org/doi/10.1021/ja00095a051) и биметаллические частицы Pd-No, Fe-Co, Fe-Ni.
