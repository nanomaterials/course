## Краткий исторический экскурс. Наноматериалы без нанотехнологий

Изучение низкоразмерных объектов (порошков, пигментов, чернил, пленок, кластеров и др.), и даже [квантово-размерных  эффектов](https://ru.wikipedia.org/wiki/Квантово-размерный_эффект) началось задолго до «нанобума». Археологические находки свидетельствуют о существовании рецептур приготовления коллоидных систем еще в античном мире, а возраст наиболее ранних биологических объектов, состоящих из строительных блоков с размерами в несколько нм, исчисляется с момента возникновения жизни на Земле.

Одними из наиболее ранних примеров получения и применения наноматериалов являются наночастицы [благородных металлов](https://ru.wikipedia.org/wiki/%D0%91%D0%BB%D0%B0%D0%B3%D0%BE%D1%80%D0%BE%D0%B4%D0%BD%D1%8B%D0%B5_%D0%BC%D0%B5%D1%82%D0%B0%D0%BB%D0%BB%D1%8B), обладающие уникальными оптическими свойствами и способностью эффективно рассеивать и поглощать электромагнитное излучение в оптическом диапазоне длин волн.

С давних времен люди применяют металлические частицы для окраски стекол. С этой целью в процессе изготовления стекол в расплавы добавляются небольшие количества примесей металлов или их оксидов. Археологические раскопки показывают, что цветные стекла по такой технологии изготовлялись в Древнем Египте во втором тысячелетии до н.э. и в Древнем Риме с I в. н.э.

*Рис. 1. Кубок Ликурга при освещении рассеянным светом снаружи (слева) и точечным источником света изнутри*

При освещении источником [белого света](https://ru.wikipedia.org/wiki/%D0%91%D0%B5%D0%BB%D1%8B%D0%B9_%D1%81%D0%B2%D0%B5%D1%82) с разных сторон кубок «светится» либо красным, либо зеленоватым цветом. Сейчас известно, что эффект обусловлен рассеянием света на содержащихся в стекле металлических наночастицах, состоящих из золота и серебра (Рис. 2).

Начиная со Средних веков и до настоящего времени металлические наночастицы широко используются в изготовлении витражей соборов. Так, металлические и полупроводниковые наночастицы применялись для окраски стекол, созданных в XIII-XIV вв. для витражей [собора Парижской Богоматери](https://ru.wikipedia.org/wiki/Собор_Парижской_Богоматери) (Рис. 3). Заметим, что, металлические наночастицы в сочетании с оксидами, входящими в состав стекла, также могут действовать как композитный [фотокалатический материал](https://ru.wikipedia.org/wiki/%D0%A4%D0%BE%D1%82%D0%BE%D0%BA%D0%B0%D1%82%D0%B0%D0%BB%D0%B8%D0%B7), препятствующий загрязнению стекол за счет разрушения органических молекул на поверхности материала.

![Северная роза собора Парижской Богоматери](Figs/North_rose_window.jpg)

*Рис. 3. Собор Парижской Богоматери: витражи северной розы (застекление 1255 г.) с фризами*

В 1612 г. флорентийский священник и мастер по стеклу Антонио Нери описал в трактате «Искусство стекла» синтез [коллоидного золота](https://ru.wikipedia.org/wiki/Коллоидное_золото). [Иоганн Кункель](https://ru.wikipedia.org/wiki/%D0%9A%D1%83%D0%BD%D0%BA%D0%B5%D0%BB%D1%8C,_%D0%98%D0%BE%D0%B3%D0%B0%D0%BD%D0%BD) в 1689 г. перевёл работу на немецкий язык и разработал технологию получения «золотого рубинового стекла» (Рис. 4).

![](Figs/Goldrubinglas.jpg)

*Рис. 4. Посуда из рубинового стекла, изготовленного Кункелем*

Секреты технологии были зановы открыты [М.В. Ломоносовым](https://ru.wikipedia.org/wiki/Ломоносов,_Михаил_Васильевич). Согласно Ломоносову для получения необходимо в обычное стекло добавить незначительное количество (0,01-0,1 %) раздробленных частиц Au диаметром 4-30 мкм.

Аналогичного эффекта можно добиться без использования частиц золота. Так, для получения рубинового стекла [звезд Московского Кремля](https://ru.wikipedia.org/wiki/%D0%9A%D1%80%D0%B5%D0%BC%D0%BB%D1%91%D0%B2%D1%81%D0%BA%D0%B8%D0%B5_%D0%B7%D0%B2%D1%91%D0%B7%D0%B4%D1%8B) (Рис. 5) в [шихту](https://ru.wikipedia.org/wiki/%D0%A8%D0%B8%D1%85%D1%82%D0%B0) добавлялись элементарный селен и соль кадмия, которые в процессе варки стекла давали микрокристаллы полупроводника AIIBVI [селенида кадмия](https://ru.wikipedia.org/wiki/%D0%A1%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B4_%D0%BA%D0%B0%D0%B4%D0%BC%D0%B8%D1%8F)  (CdSe).

![Московский Кремль](Figs/View_of_the_Moscow_Kremlin_at_night.jpg)

*Рис. 5. Вид на Московский Кремль с Пречистенской набережной. Цвет кремлевских звезд обусловлен композицией рубинового и матового стекол*

Наивную форму нанотехнологий практиковали также гончары из итальянской провинции [Умбрия](https://ru.wikipedia.org/wiki/%D0%A3%D0%BC%D0%B1%D1%80%D0%B8%D1%8F) в XV-XVI вв. В цветной глазури образцов керамики умбрийского города [Дерута](https://ru.wikipedia.org/wiki/%D0%94%D0%B5%D1%80%D1%83%D1%82%D0%B0) отражающие свойства наноразмерных металлических частиц (Cu, Ag) использовались для придания глазури необычного блеска (Рис. 6). Предметы с блестящей поверхностью получались при использовании строго регулируемой техники обжига.

![Умбрийская ваза](Figs/Deruta.jpg)

*Рис.6. Умбрийская ваза, XVI в.*

Наноматериалы без знаний о нанотехнологиях имели не только эстетическое значение. Сотрудники [Дрезденского технического университета](https://ru.wikipedia.org/wiki/Дрезденский_технический_университет), исследуя образец [дамасской стали](https://ru.wikipedia.org/wiki/Дамасская_сталь) сабли, хранящейся в Историческом музее г. Берна (Швейцария), обнаружили в структуре стали тонкие протяженные объекты нанометровых поперечных размеров. При детальном изучении в [просвечивающем электронном микроскопе](https://ru.wikipedia.org/wiki/Просвечивающий_электронный_микроскоп) высокого разрешения оказалось, что это многослойные [углеродные нанотрубки](https://ru.wikipedia.org/wiki/Углеродные_нанотрубки), заполненные [карбидом железа](https://ru.wikipedia.org/wiki/%D0%A6%D0%B5%D0%BC%D0%B5%D0%BD%D1%82%D0%B8%D1%82) $\text{Fe$_3$C}$ (Рис. 7). Таким образом, можно предположить, что повышенные прочностные свойства дамасской стали были обусловлены нанотрубками, обладающими рекордной прочностью на растяжение. 

![](Figs/Damascuc_sabre.png)

*Рис. 7. Пример нанопровода карбида железа, заключенного в многостенную углеродную нанотрубку. Обнаружено при растворении образца дамасской стали в соляной кислоте. Масштабная линейка 5 нм. Расстояние между плоскостями в проводе 0.635 нм, что соответствует межплоскостному расстоянию (010)*

Систематические научные исследования нанообъектов начались в середине XIX в., когда [М. Фарадей](https://ru.wikipedia.org/wiki/%D0%A4%D0%B0%D1%80%D0%B0%D0%B4%D0%B5%D0%B9,_%D0%9C%D0%B0%D0%B9%D0%BA%D0%BB) получил и исследовал свойства коллоидных растворов высокодисперсного золота (Рис. 8) и пленок на его основе. [Г. Ми](https://ru.wikipedia.org/wiki/Ми,_Густав) впоследствии предложил теоретические основы описания оптических свойств наночастиц, которые широко используются и сегодня. 

![](Figs/gold_sample.jpg)

*Рис. 8. Образец коллоидного золота, полученного Майклом Фарадеем в 1856 г. [Источник](http://www.rigb.org/our-history/iconic-objects/iconic-objects-list/faraday-gold-colloids)*

В первой половине XX в. [наука о коллоидах](https://ru.wikipedia.org/wiki/%D0%9A%D0%BE%D0%BB%D0%BB%D0%BE%D0%B8%D0%B4%D0%BD%D0%B0%D1%8F_%D1%85%D0%B8%D0%BC%D0%B8%D1%8F) была областью исследований лишь нескольких специализированных групп. Это побудило [В. Ф. Оствальда](https://ru.wikipedia.org/wiki/Оствальд,_Вильгельм_Фридрих) назвать свою книгу о коллоидах, вышедшую в 1915 г., «Мир обойденных величин». В то же время получили развитие  методы синтеза коллоидного золота. Для приготовления [золей](https://ru.wikipedia.org/wiki/%D0%97%D0%BE%D0%BB%D0%B8) было предложено использовать [зажигание дуги](https://ru.wikipedia.org/wiki/%D0%AD%D0%BB%D0%B5%D0%BA%D1%82%D1%80%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B0%D1%8F_%D0%B4%D1%83%D0%B3%D0%B0) между помещенным в щелочной раствор золотым электродами, пропускать газ [монооксида углерода](https://ru.wikipedia.org/wiki/%D0%9C%D0%BE%D0%BD%D0%BE%D0%BE%D0%BA%D1%81%D0%B8%D0%B4_%D1%83%D0%B3%D0%BB%D0%B5%D1%80%D0%BE%D0%B4%D0%B0) через раствор золотохлористой кислоты, а [Р. Зигмонди](https://ru.wikipedia.org/wiki/Зигмонди,_Рихард) открыл метод затравки (seeding method), и использовал формальдегид и слабую щелочь для получения золя золота из его солей. Вручение Зигмонди в 1925 г. Нобелевской премии по химии частично было обусловлено его работой с коллоидами золота.

В 1905 г. [А. Эйнштейн](https://ru.wikipedia.org/wiki/%D0%AD%D0%B9%D0%BD%D1%88%D1%82%D0%B5%D0%B9%D0%BD,_%D0%90%D0%BB%D1%8C%D0%B1%D0%B5%D1%80%D1%82) предложил количественную теорию поведения коллоидных сред, рассматривая коллоид как множество частиц, находящихся в состоянии [броуновского движения](https://ru.wikipedia.org/wiki/%D0%91%D1%80%D0%BE%D1%83%D0%BD%D0%BE%D0%B2%D1%81%D0%BA%D0%BE%D0%B5_%D0%B4%D0%B2%D0%B8%D0%B6%D0%B5%D0%BD%D0%B8%D0%B5). Его теория бьша подтверждена экспериментами [Ж. Б. Перрена](https://ru.wikipedia.org/wiki/%D0%9F%D0%B5%D1%80%D1%80%D0%B5%D0%BD,_%D0%96%D0%B0%D0%BD_%D0%91%D0%B0%D1%82%D0%B8%D1%81%D1%82), который в 1926 г. по физике за работу по дискретной природе материи был удостоен Нобелевской премии. 

К наиболее значимым последующим вехам на пуги развития нанотехнологий следует отнести работы [И. Ленгмюра](https://ru.wikipedia.org/wiki/%D0%9B%D0%B5%D0%BD%D0%B3%D0%BC%D1%8E%D1%80,_%D0%98%D1%80%D0%B2%D0%B8%D0%BD%D0%B3) (1916-1918 гг.), который экспериментально получил и исследовал моноатомные и мономолекулярные слои, то есть уже не дискретные частицы, а двухмерные материалы. В 1932 г. Ленгмюр получил Нобелевскую премию за открытия и исследования в области химии поверхности. 

Большой подъем в использовании методов коллоидной химии для получения наночастиц различных материалов произошел во второй половине XX в. благодаря возбуждению, вызванному наукой о наноматериалах, инициированному в 1959 г. лекцией Нобелевского лауреата [Р. Фейнмана](https://ru.wikipedia.org/wiki/%D0%A4%D0%B5%D0%B9%D0%BD%D0%BC%D0%B0%D0%BD,_%D0%A0%D0%B8%D1%87%D0%B0%D1%80%D0%B4_%D0%A4%D0%B8%D0%BB%D0%BB%D0%B8%D0%BF%D1%81) «Внизу полным-полно места: приглашение в новый мир физики», в которой было подчеркнуто принципиальное значение малоразмерных объектов. Фейнман акцентировал внимание на важности работ в области сжатия информации, создания миниатюрных компьютеров, дизайна материалов и устройств методами молекулярной архитектуры. Большие надежды ученый возлагал на химический синтез, отмечая, что законы физики не запрещают конструирования материалов на атомно-молекулярном уровне (вы можете прочитать приложенные к этому файлу [оригинал лекции](Feynman_en.pdf) или ее [перевод в журнале «Химия и жизнь»](Feynman_ru.pdf)).

Впервые термин «нанотехнология» употребил профессор Токийского университета [Н. Танигучи](https://ru.wikipedia.org/wiki/%D0%A2%D0%B0%D0%BD%D0%B8%D0%B3%D1%83%D1%82%D0%B8,_%D0%9D%D0%BE%D1%80%D0%B8%D0%BE) в 1974 г., который вложил в него смысл прецизионного производства изделий размером всего в несколько нанометров. С этого времени слово вошло в практику научных исследований и инженерных разработок. 

Первые средства для нанотехнологии были разработаны в Цюрихе и Рюмликоне, в швейцарских лабораториях фирмы [IBM](https://ru.wikipedia.org/wiki/IBM). В 1981 г. — [сканирующий туннельный микроскоп](https://ru.wikipedia.org/wiki/Сканирующий_туннельный_микроскоп), предназначенный для изучения атомной и молекулярной структуры поверхности проводящих веществ. Четыре года спустя, в 1986 г., его создатели [Г. Биннинг](https://ru.wikipedia.org/wiki/Бинниг,_Герд_Карл) и [Г. Рорер](https://ru.wikipedia.org/wiki/Рорер,_Генрих) получили Нобелевскую премию по физике. Однако множество трудностей, усложнявших исследование образцов в сканирующем туннельном микроскопе, побудили тех же авторов к 1986 г. разработать [атомно-силовой микроскоп](https://ru.wikipedia.org/wiki/Сканирующий_атомно-силовой_микроскоп). Эти микроскопы стали первыми воплощениями фейнмановской машины для манипуляций на атомарному уровне.

В том же 1986 г. вышла книга [Эрика Дрекслера](https://ru.wikipedia.org/wiki/Дрекслер,_Эрик) (MIT, США) «Машины созидания: пришествие эры нанотехнологии», в которой были развиты идеи Фейнмана и введены представления о молекулярных роботехнических машинах.

Параллельно концепции нанотехнологий развивалась и концепция наноматериалов как материалов со свойствами, отличными от макроскопических веществ. Впервые эти идеи были сформулированы в 1981 г. Г. Глейтером применительно к металлическим материалам. Главная роль была отведена поверхностям раздела (границам зерен) как фактору, позволяющему существенно изменить свойства твердых тел путем модификации структуры и электронного строения. Глейтером был предложен метод получения ультадисперсных порошков путем испарения-конденсации с последующей вакуумной консолидацией при высоких давлениях. Первые работы в этом направлении дали практический успех — полученные в 1983 г. по методу Глейтера нанокристаллические образцы никеля обладали вдвое большей твердостью, чем обычный поликристаллический никель.
