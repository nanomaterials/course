# Граничные эффекты. Влияние доли поверхностных атомов

Если при уменьшении объема вещества по одной, двум или трем координатам до размеров нанометрового масштаба возникает новое качество, или это качество возникает в композиции из таких объектов, то такие образования относят к [наноматериалам](https://ru.wikipedia.org/wiki/Наноматериал), а технологии их получения и дальнейшую работу с ними — к нанотехнологиям. Уникальные свойства наноматериалов определяются их структурой, в частности, структурой поверхности и границ раздела. Роль, которую играют размер и структура [наночастиц](https://ru.wikipedia.org/wiki/%D0%9D%D0%B0%D0%BD%D0%BE%D1%87%D0%B0%D1%81%D1%82%D0%B8%D1%86%D0%B0), сравнима с ролью химического состава вещества.

Таким образом, сущность нанотехнологии состоит в возможности работать на атомарном и молекулярном уровнях в масштабе линейных размеров 1–100 нм, для того, чтобы создавать и использовать материалы и устройства, имеющие новые свойства и функции. Диапазон размеров 1–100 нм является очевидно условным и скорее характеризует порядок величин, чем сами границы диапазона. Например, [фуллерен](https://ru.wikipedia.org/wiki/%D0%A4%D1%83%D0%BB%D0%BB%D0%B5%D1%80%D0%B5%D0%BD) C$_{60}$ относят и к молекулам, и к наноматериалам. В одних случаях интервал оказывается существенно уже, в других —  выходит за 100 нм. Нижняя граница в 1 нм обусловлена тем, что для объектов меньше 1 нм, таких как атомы и молекулы, научные дисциплины начали развиваться сравнительно давно и в большей мере относятся к областям [квантовой физики](https://ru.wikipedia.org/wiki/%D0%9A%D0%B2%D0%B0%D0%BD%D1%82%D0%BE%D0%B2%D0%B0%D1%8F_%D1%84%D0%B8%D0%B7%D0%B8%D0%BA%D0%B0) и [квантовой химии](https://ru.wikipedia.org/wiki/%D0%9A%D0%B2%D0%B0%D0%BD%D1%82%D0%BE%D0%B2%D0%B0%D1%8F_%D1%85%D0%B8%D0%BC%D0%B8%D1%8F).

Наноматериалы характеризуются важными особенностями, отличающими их от макроскопических образцов тех же веществ:
1. Сложная внутренняя организация наноматериалов существенно отличается от структуры крупных частиц (в т. ч. по типу структуры кристаллической решетки и ее симметрии).
2. Сильная зависимость свойств от линейных размеров. К числу этих свойств относятся как физические свойства (тепловые, механические, электрические, магнитные), так и химические (каталитическая активность, растворимость). В ряде случаев свойства наночастиц имеют квантово-размерный характер, т.е. наночастицы можно рассматривать как молекулы с дискретным строеним энергетических уровней.

Наблюдаемые эффекты можно условно разделить на *граничные* – связанные с увеличением доли поверхностных атомов либо межзеренных границ и *размерные* – обусловленные изменением общего количества атомов в составе единичной структуры. Это условное разделение показывает какой из этих аспектов играет более значимую роль. В этом разделе мы подробнее остановимся на граничных эффектах.

##  Доля поверхностных атомов частицы материала

Представим однородный твердый материал сферической формы, имеющий макроскопические линейные размеры порядка единиц мм. Большинство свойств материала будет определяться химическим составом и кристаллической структурой. Поверхностные атомы составляют незначительную долю от общего числа атомов и, следовательно, играют незначительную роль в свойствах материала. 

При рассмотрении на атомном уровне поверхность твердого тела представляет собой зону нескомпенсированных связей. Особенностью наночастиц является то, что доля поверхностных атомов, ионов или молекул возрастает в сравнении с долей элементов, находящихся внутри частицы. Определим долю поверхностных атомов $w_s$ как отношение числа атомов, находящихся на поверхности $N_s$ к общему числу атомов $N$:

$$
w_s = \dfrac{N_s}{N}.
$$

Считая, что атомы на поверхности занимают тот же объем, что внутри структуры, можно представить долю поверхностных атомов через отношение объема поверхностного слоя толщиной в диаметр атома ($2r$) к объему тела $V$:

$$
w_s = \dfrac{V_s}{V} \approx \dfrac{S \cdot 2r}{V} = \dfrac{4\pi R^2 2r}{\dfrac{4}{3}\pi R^3} = \dfrac{6r}{R}.
$$

Несколько более точно можно рассмотреть эту задачу, разбив объем на соответствующий поверхностному слою $V_s$ и внутреннему объему $V_v$:

$$
w_s = \dfrac{V_s}{V} = \dfrac{V-V_v}{V} = \dfrac{R^3-(R-2r)^3}{R^3} = \dfrac{6R^2r-12Rr^2+8r^3}{R^3}.
$$

При сравнении выражений видно, что первое решение является частным случаем последнего для $R >> r$.

```python
%matplotlib inline
import numpy as np

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 14})


def w_s_0(r, R):
    '''Менее точное, но наглядное выражение'''
    return 6*r/R

def w_s(r, R):
    '''Более точный вариант'''
    return (6*R**2*r - 12*R*r**2 + 8*r**3)/R**3

r_Au = 136e-12   # ковалентный радиус атома золота
R = np.linspace(1, 10)*1e-9

plt.plot(R/1e-9, w_s_0(r_Au, R), label='$\dfrac{6r}{R}$')
plt.plot(R/1e-9, w_s(r_Au, R), label='$\dfrac{6R^2r-12Rr^2+8r^3}{R^3}$')
plt.xlabel('$R$, нм')
plt.ylabel('$w_s$')
plt.legend()
plt.tight_layout()
plt.savefig('radius.png')
plt.show()
```

*Рис. 1. Зависимость доли поверхностных атомов от радиуса золотой наночастицы по приведенным выражениям*

Из приведенного построения можно видеть, что уже для наночастиц с диаметром менее 2 нм доля поверхностных атомов составляет более 50%.

### Эффект изменения постоянной кристаллической решетки

Важным свойстом границ наноразмерных структур является конечная, крайне малая величина радиуса [кривизны поверхности](https://ru.wikipedia.org/wiki/%D0%9A%D1%80%D0%B8%D0%B2%D0%B8%D0%B7%D0%BD%D0%B0), в отличие от бесконечной кривизны граней монокристалла. Рассмотрим сферическую частицу диаметром $2R$. Искривление поверхности приводит к [избыточному давлению](https://ru.wikipedia.org/wiki/Капиллярное_давление) внутри материала частицы. Найти величину этого давления можно из термодинамики через описание изменения [энергии Гиббса](https://ru.wikipedia.org/wiki/%D0%AD%D0%BD%D0%B5%D1%80%D0%B3%D0%B8%D1%8F_%D0%93%D0%B8%D0%B1%D0%B1%D1%81%D0%B0) $dG$. Энергия затрачивается на работу изобарического расширения и работу на изменение поверхности частицы:

$$dG = p dV + \gamma dS,$$

где $dV$ — изменение объема, соответствующее изменению поверхности на величину $dS$, $\gamma$ — [поверхностное натяжение](https://ru.wikipedia.org/wiki/%D0%9F%D0%BE%D0%B2%D0%B5%D1%80%D1%85%D0%BD%D0%BE%D1%81%D1%82%D0%BD%D0%BE%D0%B5_%D0%BD%D0%B0%D1%82%D1%8F%D0%B6%D0%B5%D0%BD%D0%B8%D0%B5).

В условиях термодинамического равновесия $dG = 0$:

$$p dV + \gamma dS = 0,$$
$$4\pi p R^2 dR + 8 \pi \gamma R  dR = 0,$$

где $R$ — радиус частицы. Таким образом, в случае сферы избыточное давление равно:

$$p = -\dfrac{2\gamma}{R}.$$

Это уравнение называют уравнением Лапласа. Отрицательный знак показывает, что внутри частицы давление выше, чем в окружающей среде. 

В общем случае поверхностная энергия задается тензором напряжений, однако в изотропном приближении малой частицы можно считать эту величину скаляром, равным поверхностному натяжению. С другой стороны, существует понятие [сжимаемости вещества](https://ru.wikipedia.org/wiki/Сжимаемость). Соответствующий коэффициент всестороннего сжатия определяется выражением

$$\beta = \dfrac{1}{V}\dfrac{\Delta V}{p},$$

$$\beta = \dfrac{1}{a^3}\dfrac{a^3 - (a-\Delta a)^3}{p} \approx \dfrac{\Delta a}{a}\dfrac{3}{p}.$$

Выразим относительное изменение постоянной решетки, учитывая вышеприведенное выражение для избыточного давления:

$$\dfrac{\Delta a}{a} = -\dfrac{2}{3}\dfrac{\beta \gamma}{R}.$$

Относительно справочных данных для коэффициента сжимаемости удобнее рассматривать обратную величину — [объемный модуль упругости](https://ru.wikipedia.org/wiki/Объёмный_модуль_упругости) $B = \beta^{-1}$, тогда:

$$\dfrac{\Delta a}{a} = -\dfrac{2}{3}\dfrac{\gamma}{BR}.$$

Таким образом, относительное изменение постоянной решетки растет с уменьшением размера наночастицы. Абсолютные размеры наночастиц в результате всестороннего сжатия оказываются меньше ожидаемых без учета избыточного давления.

```python
import pandas as pd

metals = pd.read_csv('../Data/metals.csv')

R = np.linspace(1, 10)*1e-9

for i, material in metals.iterrows():
    gamma = material['Поверхностное натяжение, Дж/м2']
    K = material['Объемный модуль упругости, ГПа']*1e9
    label = material['Материал']
    plt.plot(R/1e-9, -2/3*gamma/(K*R)*100, label=label)

plt.xlabel('$R$, нм')
plt.ylabel('$\dfrac{\Delta a}{a}$, %')
plt.legend()
plt.tight_layout()
plt.savefig('Figs/delta_a.png')
plt.show()
```

*Рис. 2. Зависимость относительного изменения (уменьшения) постоянной решетки в зависимости от радиуса наночастиц*

**Пример**. При рентгеноструктурном анализе параметров порошков BaTiO$_3$, полученного методами гидротермального синтеза обнаружено, что соотношение длин осей $c/a$, характерное для тетрагональной модификации, снижалось, что говорит о повышении симметричности при уменьшении размера частиц ниже 200 нм. Это можно объяснить воздействием на частицы сил сжатия, возникающих в результате собственного поверхностного натяжения.

### Эффект изменения длины связей в приповерхностном слое

Рассмотренный выше эффект влияния избыточного давления имеет усредненный характер, а расчетные формулы выведены из макроскопических предположений и термодинамических и механических характеристик объемных материалов. Поэтому для наночастиц они не вполне справедливы.

В то же время для приповерхностного слоя атомов зачастую наблюдается еще большее изменение:

- **Пример 1**. Методом дифракции медленных электронов установлено $\sim$10%-ое уменьшение первого межслоевого $(10-10)$ расстояния на поверхности частиц Ru, Co, Re.
- **Пример 2**. Межслоевое расстояние между первым и вторым слоем (111) поверхности алмаза на $30\%$ меньше, чем $d_{111}$ в объеме вещества.
- **Пример 3**. С помощью дифракции очень медленных электронов было обнаружено 4-12%-ое сокращение связи O–Cu на  поверхности $(001)$ кристалла и $10\%$-ое сокращение на поверхности $(110)$.

Уменьшение длины связей на поверхности уменьшает удельную поверхностную энергию системы. Поэтому кроме эффекта избыточного давления, воздействующего на всю частицу в целом, необходимо учитывать «обрыв» [периодического потенциала](https://ru.wikipedia.org/wiki/%D0%A7%D0%B0%D1%81%D1%82%D0%B8%D1%86%D0%B0_%D0%B2_%D0%BF%D0%B5%D1%80%D0%B8%D0%BE%D0%B4%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%BE%D0%BC_%D0%BF%D0%BE%D1%82%D0%B5%D0%BD%D1%86%D0%B8%D0%B0%D0%BB%D0%B5), создаваемого кристаллической решеткой у поверхности. Для поверхностных атомов характерны:

1. Уменьшение [координационного числа](https://ru.wikipedia.org/wiki/%D0%9A%D0%BE%D0%BE%D1%80%D0%B4%D0%B8%D0%BD%D0%B0%D1%86%D0%B8%D0%BE%D0%BD%D0%BD%D0%BE%D0%B5_%D1%87%D0%B8%D1%81%D0%BB%D0%BE) атомов ($z$).
2. Возникновение поверхностного [потенциального барьера](https://ru.wikipedia.org/wiki/%D0%9F%D0%BE%D1%82%D0%B5%D0%BD%D1%86%D0%B8%D0%B0%D0%BB%D1%8C%D0%BD%D1%8B%D0%B9_%D0%B1%D0%B0%D1%80%D1%8C%D0%B5%D1%80).

Уменьшение $z$ поверхностных атомов приводит к возникновению локально-избыточной [электронной плотности](https://ru.wikipedia.org/wiki/%D0%AD%D0%BB%D0%B5%D0%BA%D1%82%D1%80%D0%BE%D0%BD%D0%BD%D0%B0%D1%8F_%D0%BF%D0%BB%D0%BE%D1%82%D0%BD%D0%BE%D1%81%D1%82%D1%8C) и сокращению длин связей с остальными атомами. Одним из подходов к описанию изменения длины связей в плотноупакованных решетках является приближение В. Гольдшмидта:

$$
\frac{\Delta a_1} {a} = \dfrac{1 - e^\frac{12-z}{8z} } {1 + e^\frac{12-z}{8z} }.
$$

```python
def dd(z):
    p = (12-z)/(8*z)
    return (1-np.exp(p))/(1+np.exp(p))

z = np.arange(12, 1, -1)
plt.plot(z, dd(z)*100, 'o')
plt.xlabel('$z$')
plt.ylabel('$\dfrac{\Delta a_1} {a}$, %')
plt.tight_layout()
plt.savefig('Figs/delta_a1.png')
plt.show()
```

*Рис. 3. Зависимость относительного изменения длины связей приповерхностного слоя от координационного числа в соответствии с приближением Гольдшмидта*

Таким образом, сокращение длины связей поверхностных атомов $\frac{\Delta a_1} {a}$ может иметь существенно большие величины, чем усредненное изменение $\frac{\Delta a} {a}$, создаваемое за счет изменения кривизны поверхности $R$. Аналогичное сокращение межатомных расстояний во внешних атомных слоях наблюдается для большинства металлов, хотя в литературе встречаются и противоположные примеры.

### Эффект уменьшения температуры плавления

В связи с тем, что поверхностные атомы обычно более подвижны, температура плавления в наноматериалах $T_\text{пл}$ обычно уменьшается по сравнению с температурой плавления вещества в макроскопическом состоянии. В 1954 г. японский ученый М. Такаге методом дифракции электронов обнаружил эффект уменьшения $T_\text{пл}$ при уменьшении толщин пленок легкоплавких металлов (Pb, Sn и Bi) от 100 до 1 нм (Рис. 4).

![](Figs/Takage-diffraction.png)
*Рис. 4. Дифракционные картины, наблюдавшиеся Такаги на образцах  при нагреве тонких пленок до (снизу) и после температуры плавления*

Уравнение, связывающее температуру плавления с размером наночастиц:

$$
T_\text{пл}(R) = T_\text{пл}(\infty)\left(1 - \frac{2 \sigma}{R}\frac{M}{\rho_{s}\Delta H_\text{пл}}\right),
$$

где $T_\text{пл}(\infty)$ — температура плавления объемного материала, $\sigma$ — поверхностное натяжение между жидкой и твердой фазами, $\Delta H_\text{пл}$ — удельная теплота плавления, $M$ — молярная масса вещества, $\rho_\text{тв}$ — плотность твердого вещества, $R$ — радиус частиц.

Изменение температуры может составлять несколько сотен К. Понижение температуры тем больше, чем выше ее значение в недиспергированном состоянии и величина поверхностного натяжения.

**Пример**. Температура плавления золота в виде объемного образца составляет 1336 К, однако эта температура начинает заметно снижаться при размерах частиц меньше 20 нм и уменьшается более чем на 500 К для наночастиц диаметром 2 нм. Покажем это на примере расчета:

```python
Tsl = 1336     # K
sigma = 1.102  # Н/м
rho = 19320    # кг/см3
Hl = 12500     # Дж/моль
M = 0.197      # кг/моль

def T(R):
    Tsl_R = Tsl*(1 - 2*sigma/R*M/rho/Hl)
    return Tsl_R

R = np.linspace(5e-9, 1e-7)

plt.plot(R/1e-9, T(R))
plt.xlabel('$R$, нм')
plt.ylabel('$T_{пл}$, К')
plt.tight_layout()
plt.savefig('Figs/melting.png')
plt.show()
```

*Рис. 5. Зависимость температуры плавления от линейных размеров частиц материала*

Эффект уменьшения температуры плавления ультрадисперсных частиц имеет важный практический аспект и активно используется для повышения эффективности спекания керамических материалов.

2. В результате коллоидного синтеза получены наночастицы золота. Из результатов просвечивающей электронной микроскопии определено, что средний диаметр частиц составляет 10 нм. Оцените в процентах соответствующую долю поверхностных атомов. Радиус атомов золота примите равным 136 пм.

3. Во сколько раз капиллярное давление капли ртути радиусом 10 нм превышает атмосферное давление, если поверхностное натяжение ртути при комнатной температуре составляет 486,5 мДж/м²?

4. В соответствии с указанной формулой определите темпетратуру плавления (в градусах Кельвина) наночастицы золота радиусом 5 нм. Температура плавления золота в виде объемного образца составляет 1336 К, поверхностное натяжение между твердой и жидкой фазами равно 1.102  Н/м, плотность золота равна 19320 кг/см3, энтальпия плавления составляет 12.5 кДж/моль, молярную массу определите из таблицы Менделеева.

5. Длина ребра кубической частицы с простой кубической решеткой равна 2.1 нм (длина ребра определяется как расстояние между центрами ближайших угловых атомов). Определите долю позиций, в которых атомы имеют 3, 4 и 5 соседей. Постоянная решетки материала равна 2.1 Å. Изменением межатомных расстояний пренебречь.
