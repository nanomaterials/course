# Лабораторная работа 2. Исследование оптических свойств металлических и полупроводниковых нанокристаллов

**Цель** – исследование оптических свойств металлических и полупроводниковых наночастиц, проявляющих соответственно локализованных поверхностный плазмонный резонанс и квантово-размерные эффекты, методами спектроскопии поглощения и фотолюминесценции.

## Основные понятия и определения

Так как основные замечания, касающиеся технологии синтеза и физических свойств металлических и полупроводниковых нанокристаллов, проявляющих соответственно локализованный плазмонный поверхностный резонанс и квантово-размерные эффекты, описаны в теоретических разделах, не будем их дублировать.

Обратим внимание на ключевые идеи экспериментальных методик, испольщующихся в данной лабораторной работе.

Спектром отражения в случае коллоидных растворов в силу малой величины коэффициентов отражения можно пренебречь. Поэтому оптическая плотность является удобной величиной для дальнейшего построения спектров поглощения. В общем случае пропускание света каким-либо материалом описывается законом Бугера:

$$T = \dfrac{I}{I_0} = e^{-\alpha l},$$

где $T$ – коэффициент пропускания, $I$ и $I_0$ – интенсивности прошедшего и падающего излучения соответственно, $\alpha$ – показатель поглощения, $l$ – длина оптического пути (при измерении спектров растворов – ширина кюветы).

Часто для практических целей закон Бугера выражают в ввиде степенной зависимости:

$$T = \dfrac{I}{I_0} = 10^{-\alpha'l},$$

где $\alpha'$ имеет смысл показателя поглощения при представлении $T$ через десятичный логарифм. В таком случае удобно пользоваться величиной оптической плотности $A$:

$$A = \alpha'l = -\lg T.$$

Оптическая плотность, равная $1$, соответствует $10%$ пропускания, плотность $2$ соответствует $1%$ пропускания и т. д. Использование оптической плотности облегчает расчеты концентраций и выбор оптического пути кювет для измерений.

Показатель поглощения $k$ характеризует свойства вещества и зависит от
длины волны $\lambda$ поглощаемого света. Эта зависимость называется *спектром
поглощения* вещества. Спектр поглощения полупроводника вблизи края поглощения обычно имеет нелинейную форму.

Для того чтобы однозначным образом определить энергетический зазор, необходимо линеаризовать зависимость вблизи края поглощения, для чего обычно используется соотношение Тауца, описывающее поведение спектра вблизи края поглощения:

$$\alpha h \nu = B(h \nu - E^*_g)^n,$$

где $B$ – константа; $n$ – постоянная, зависящая от типа оптических переходов
(для прямых разрешенных переходов $n = 1/2$, для непрямых разрешенных $n = 2$). Таким образом, для определения энергетического зазора полупроводника, например в случае прямых разрешенных переходов, к линеаризованному участку графика в координатах Тауца ($\alpha h \nu)^2$ от $h \nu$ строят касательную и принимают точку ее пересечения с осью абсцисс за значение энергетического зазора.

Однако стоит заметить, что в случае квантовых точек рассмотрение координат Тауца стоит использовать с осторожностью. Для монодисперсных наночастиц с размером менее $r_B$, то есть при ярком проявлении квантоворазмерных эффектов, на участке вблизи края поглощения проявляются особенности, связанные с экситонным спектром.

## Описание установки для измерения спектров поглощения

Исследование спектров поглощения коллоидных растворов наночастиц проводится на однолучевом спектрофотометре ПЭ-5400УФ. Спектральный диапазон прибора 190-1000 нм.

Для проведения измерений исследуемые коллоидные растворы помещаются в кварцевые кюветы, прозрачные в спектральном диапазоне прибора. Кюветы устанавливаются в четырехпозиционный держатель.

Принцип действия фотометра основан на сравнении светового потока $\Phi_0$, прошедшего через раствор сравнения (контрольный раствор, по отношению к которому производится измерение), и светового потока $\Phi$, прошедшего через исследуемую среду. Световые потоки $\Phi_0$ и $\Phi$ преобразуются фотоприемником в электрические сигналы $I_0$ и $I$. Также измеряется $I_Т$ – сигнал неосвещенного приемника. По уровням этих сигналов микропроцессором
спектрофотометра на каждой длине волны рассчитывается и отображается на дисплее результат измерения в виде коэффициента пропускания $T$ в процентах:

$$T = \dfrac{\Phi}{\Phi_0} \cdot 100\% = \dfrac{I - I_T}{I_0 - I_T} \cdot 100\% $$

или оптической плотности $A$:

$$A = \lg\dfrac{\Phi_0}{\Phi}= \lg\dfrac{100\%}{T}.$$

## Описание установки для измерения спектров фотолюминесценции

Спектры фотолюминесценции образцов исследуются при помощи спектрометра на основе монохроматора МДР-206. Для проведения измерений образец в кварцевой кювете помещается в держатель и подвергается воздействию возбуждающего излучения полупроводникового лазера с длиной волны 405 нм. Длина волны источника излучения должна соответствовать полосе поглощения исследуемых образцов. Прибор последовательно выделяет из спектра фотолюминесценции образца монохроматическое излучение в заданном диапазоне длин волн. Световой поток регистрируется фотоприемным устройством на основе кремниевого фотодиода (диапазон измерений 190-1100 нм), расположенного на выходной щели монохроматора.

## Исследование спектров поглощения наночастиц

Измерить спектры поглощения образцов, для чего последовательно выполнить следующие действия:
1. По окончании прогрева спектрофотометра запустить программу измерения спектров SC5400.
2. Задать диапазон и шаг сканирования, нажав кнопку «Параметры сканирования» (при подведении курсора к кнопкам программа отображает соответствующие подсказки). В открывшемся окне в графах поля «Длина волны» задать  по указанию преподавателя шаг измерений, начальную и конечную длины волн.
3. Поместить кювету с раствором сравнения (дисперсионная среда образца) в держатель образцов.
4. Нажать кнопку «Сканировать нулевую линию».
5. По окончании сканирования нулевой линии открыть отделение для образцов, закрепить исследуемый образец в держателе спектрофотометра, закрыть крышку.
6. Перемещая ручку держателя образцов, поставить образец в позицию для измерений и сканировать его спектр, нажав кнопку «Сканировать образец». По окончании измерения спектра нажать «Файл», далее «Экспорт» и экспортировать образец в xls-файл на сменный носитель под названием.
7. Записать в протокол материал обраца. Для металлических наночастиц записать длину волны плазмонного резонанса $\lambda_{пл}$. Для полупроводниковых наночастиц – длину волны первого экситонного пика $\lambda_{экс}$.
8.  Повторить измерения для остальных образцов.  Если дисперсионная среда образцов не одна и та же, то необходимо сначала сканировать нулевую линию дисперсионной среды.
9. Сохранить спектры на личном носителе информации.

## Исследование спектров фотолюминесценции полупроводниковых наночастиц

Измерить спектры фотолюминесценции образцов, для чего последовательно выполнить следующие действия:

1. Включить монохроматор в сеть питания. Запустить программу управления монохроматором МДР-206. Для того чтобы установить механизм сканирования монохроматора в первоначальное положение, в пункте меню «Прибор» запустить команду «Привязаться к нулю».
2. На панели инструментов нажать кнопку «Установка параметров сканирования» и установить режим (непрерывный), диапазон и шаг сканирования, указанные преподавателем.
3. Поместить кювету с образцом в держатель, включить источник возбуждающего излучения. На панели инструментов нажать кнопку «Запуск измерений». Измеренный спектр сохранить на носителе информации, нажав накнопку «Сохранить» в окне «Программируемый эксперимент». Запишите в протокол длину волны максимума люминесценции $\lambda_{макс}$.
4. Повторить измерения для остальных образцов. 
5. Выделив спектры в правой панели программы, последовательно сохранить спектры на личном носителе информации в виде txt-файлов.
